## Question 3

Is the following statement True or False?

> Average age of female athletes who attended the olympic games after 1990 has raised when compared to the era before then.

```python
# Your answer goes here
subset_before_1990 = athlete_events[(athlete_events['Sex'] == 'F') & (athlete_events['Year'] < 1990)]
subset_before_1990['Age'].mean()
```

```python
# Your answer goes here
subset_after_1990 = athlete_events[(athlete_events['Sex'] == 'F') & (athlete_events['Year'] >= 1990)]
subset_after_1990['Age'].mean()
```

## Question 4

How many Gold medals were given to men from 1970 to 2000 (including both years)?

```python
# Your answer goes here
men_gold = athlete_events[(athlete_events['Sex'] == 'M') & (athlete_events['Year'] >= 1970) & (athlete_events['Year'] <= 2000) & (athlete_events['Medal'] == 'Gold')]
men_gold['Medal'].count()
```
