b) Fixing Inconsistencies in Date Column

```python
#Data Cleaning for Date Under a Function

import pandas as pd

def clean_date_column(df, colname, dayfirst=True):
    """
    Clean and standardize a date column:
    - Strips spaces
    - Replaces separators with '/'
    - Converts to datetime64[ns]
    - Removes rows with dates in the future
    - Creates Year, Month, Day columns
    """
    # 1. strip and normalize separators
    df[colname] = df[colname].astype(str).str.strip()
    df[colname] = df[colname].str.replace(r'[-.]', '/', regex=True)

    # 2. convert to datetime
    df[colname] = pd.to_datetime(df[colname],
                                 dayfirst=dayfirst,
                                 errors='coerce')

    # 3. remove future dates
    today = pd.Timestamp.today().normalize()
    df = df[df[colname] <= today]

    # 4. add clean string version and parts
    clean_name = f"{colname}_clean"
    df[clean_name] = df[colname].dt.strftime('%m/%d/%Y')
    df[f"{colname}_Year"] = df[colname].dt.year
    df[f"{colname}_Month"] = df[colname].dt.month
    df[f"{colname}_Day"] = df[colname].dt.day
    df[f"{colname}_MonthYear"] = df[colname].dt.to_period('M').astype(str)

    return df

# applying to the table
T_crime_unique = clean_date_column(T_crime_unique, 'Date Rptd', dayfirst=True)
T_crime_unique = clean_date_column(T_crime_unique, 'DATE OCC', dayfirst=True)
```

g) Consolidated Yearly Tables into one Dataset with a Year + Month Column, Removed Preamble/Titles, and Dropped Unnecessary Rows Within Employment Data Set

```python
path= '/content/drive/MyDrive/Colab Notebooks/780 project/1990-2025 employment & unemployment.csv'

# Read Entire CSV With No Header (Includes Title Rows)
raw = pd.read_csv(path, header=None)

# Normalize First Column For Matching (String Type + Trim Whitespace)
s = raw[0].astype(str).str.strip()

# Locate First True Data Row (First Cell That Is Exactly A 4-Digit Year)
first_year_idx = s.str.fullmatch(r"\d{4}").idxmax()

# Keep Rows From First Year Down (Drop Cover/Preamble Above The Table)
raw = raw.iloc[first_year_idx:]

# Recompute Cleaned First Column After Slicing (Stay Aligned With New Frame)
s = raw[0].astype(str).str.strip()

# Build Year Labels From 4-Digit Rows And Forward-Fill To Month Rows
# This Copies The Year Down So Each Month Has The Correct Year
year = pd.to_numeric(s.where(s.str.fullmatch(r"\d{4}")), errors="coerce").ffill()

# Map Month Names To Numbers
MONTHS = {
    'january':1,'february':2,'march':3,'april':4,'may':5,'june':6,
    'july':7,'august':8,'september':9,'october':10,'november':11,'december':12
}

# Select Monthly Rows And Exclude Annual Averages (Keep Only True Months)
mask = s.str.lower().isin(MONTHS) & ~s.str.lower().str.startswith("annual averages")

# Derive Output Column Names From First-Year Row (Use Columns 1–4 As Headers)
h1, h2, h3, h4 = raw.loc[first_year_idx, 1:4].astype(str).str.strip().tolist()

# Convert Month Names To Numbers For Monthly Rows
mnum = s[mask].str.lower().map(MONTHS).astype(int)

# Build YYYY-MM Keys
ym = year[mask].astype(int).astype(str) + "-" + mnum.astype(str).str.zfill(2)

# Assemble Tidy Table: One Row Per Month With Four Series Columns
LA_out = pd.DataFrame({
    "ym": ym,
    h1: raw.loc[mask, 1],
    h2: raw.loc[mask, 2],
    h3: raw.loc[mask, 3],
    h4: raw.loc[mask, 4],
}).reset_index(drop=True)

LA_out[h1] = pd.to_numeric(LA_out[h1].astype(str).str.replace(",", ""), errors="coerce")
LA_out[h2] = pd.to_numeric(LA_out[h2].astype(str).str.replace(",", ""), errors="coerce")
LA_out[h3] = pd.to_numeric(LA_out[h3].astype(str).str.replace(",", ""), errors="coerce")
LA_out[h4] = pd.to_numeric(LA_out[h4].astype(str).str.rstrip("%").str.replace(",", ""), errors="coerce") / 100

# Preview First 5 Rows (Check)
print(LA_out.shape)
print(LA_out.head())
```

### 5.3 Top 15 Crimes by Victim Sex


#### 5.3.1 Victim's Gender affected by the Total Crimes
What are the top 15 crimes during these years, and does the victim's gender distribution are different across different crime types? To be more specific, which crimes have the highest proportion of female victims?

```python
#Grouping the T_crimes by Crime Description for the graph below

TOP_N = 15
crime_counts_orig = (
    T_crime.groupby('Crm Cd Desc', dropna=False)
           .size()
           .reset_index(name='count')
)
top_15_crimes = (
    crime_counts_orig.sort_values('count', ascending=False)
                     .head(TOP_N)['Crm Cd Desc']
)
# Counts by sex for ONLY those top crimes (keep Unknowns together)
crime_sex_counts = (
    T_crime[T_crime['Crm Cd Desc'].isin(top_15_crimes)]
      .assign(Vict_Sex=T_crime['Vict Sex'].fillna('Unknown'))
      .groupby(['Crm Cd Desc', 'Vict_Sex'], dropna=False)
      .size()
      .reset_index(name='count')
)
# Order y-axis by total counts so every facet shares the same order
y_order = (crime_sex_counts.groupby('Crm Cd Desc')['count']
           .sum().sort_values(ascending=False).index.tolist())
plt.figure(figsize=(12,8))
```

```python
# Pivot so Each Sex is a Column (M/F/Unknown as Separate Columns per Crime Type)
pivot_data = crime_sex_counts.pivot(index='Crm Cd Desc',
                                    columns='Vict_Sex',
                                    values='count').fillna(0)

# Reorder Rows According to a Pre-Computed Order (y_order, from earlier step)
pivot_data = pivot_data.loc[y_order]

# Ensure All Three Columns Exist, Even if One Category is Missing in the Data
for col in ['M','F','Unknown']:
    if col not in pivot_data.columns:
        pivot_data[col] = 0

# Compute Total Victims Per Crime Type
totals = pivot_data[['M','F','Unknown']].sum(axis=1)

# Compute Share (Percentage of Each Sex Per Crime Type)
shares = pivot_data[['M','F','Unknown']].div(totals, axis=0).fillna(0)
plt.figure(figsize=(12, 8))

# Define Fixed Colors for Each Sex Using Your Color Scheme
colors = {'M': COLORS['primary_blue'], 'F': COLORS['accent_gold'], 'Unknown': COLORS['light_blue']}

# Initialize "Left" Positions for Stacking (Start at 0 for All Rows)
left = np.zeros(len(pivot_data))
ax = plt.gca()

# Y Positions of Each Crime Type (Rows)
ypos = np.arange(len(pivot_data.index))

# Loop over each sex and plot its stacked segment
for sex in ['M', 'F', 'Unknown']:
    vals = pivot_data[sex].values   # absolute counts for this sex

    # Draw horizontal bars stacked on "left"
    ax.barh(pivot_data.index, vals, left=left,
            label={'M':'Male','F':'Female','Unknown':'Unknown'}[sex],
            color=colors[sex])

    # Convert fractions to percentages
    pcts = (shares[sex].values * 100.0)

    # Loop row by row: l = starting x, w = width (count), p = percent
    for i, (l, w, p) in enumerate(zip(left, vals, pcts)):
        if w <= 0:    # skip if no bar to draw
            continue
        x_center = l + w/2.0   # midpoint of the bar segment

        # If segment is big enough, draw percentage INSIDE the bar
        if p >= 4:  # threshold: at least 4% of total
            ax.text(x_center, i, f"{p:.1f}%", ha='center', va='center',
                    fontsize=9, color='white', fontweight='bold')
        else:
            # Otherwise draw just outside the bar so text is visible
            ax.text(l + w + totals.max()*0.006, i, f"{p:.1f}%", ha='left', va='center',
                    fontsize=8, color='black')
    # Update "left" so the next segment stacks on top of this one
    left += vals

# Extend x-axis a little to make room for labels placed outside bars
ax.set_xlim(0, totals.max() * 1.12)

#Plot
plt.title(f"Top {TOP_N} Crimes by Victim Sex (Stacked with % Labels)", fontsize=16, weight='bold')
plt.xlabel("Number of Records")
plt.ylabel("Crime Type")
plt.legend(title="Victim Sex")
plt.tight_layout()
plt.show()
```

Victim gender distribution varies notably across these categories. Male victims dominate in crimes like shoplifting -assault with a deadly weapon (72.6% male), and robbery (69.8% male). In contrast, female victims are most heavily represented in intimate partner – simple assault (78.7% female), followed by theft of identity (52.8% female) and battery – simple assault (50.6% female). #### 5.4.3 Top 10 Positive and Negative Correlations between Crime Types and Unemployment Rate
Which crime types show the strongest relationship with unemployment rates, and what does this reveal about broader social or economic patterns in Los Angeles?

```python
# Aggregate total crimes per Area, Crime Type, and Month
crime_counts = (
    merged_df2.groupby(['AREA NAME', 'Crm Cd Desc', 'ym'], as_index=False)
    .size()
    .rename(columns={'size': 'Crime_Count'})
)

crime_unemp = crime_counts.merge(
    merged_df2[['AREA NAME', 'ym', 'Unemp. Rate']].drop_duplicates(),
    on=['AREA NAME', 'ym'],
    how='left'
)


# Correlation of crime vs unemployment per area
corrs = (
    crime_unemp.groupby('Crm Cd Desc')
    .apply(lambda x: x['Crime_Count'].corr(x['Unemp. Rate']))
    .dropna()
    .reset_index(name='Correlation')
    .sort_values(by='Correlation', ascending=False)
)
```

```python

top_corrs = corrs.head(10)
bottom_corrs = corrs.tail(10)
subset_corrs = pd.concat([top_corrs, bottom_corrs])

plt.figure(figsize=(8,6))
sns.barplot(data=subset_corrs, y='Crm Cd Desc', x='Correlation', palette="coolwarm")
plt.title('Correlation Between Unemployment Rate and Crime Types (2018–2023)', fontsize=14, fontweight='bold')
plt.xlabel('Correlation Coefficient')
plt.ylabel('')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

This visualization examines how different crime types correlate with unemployment rates from 2018 to 2023. The results show minimal overall correlation, with only a few crimes - such as purse snatching and child neglect - displaying mild positive associations. Conversely, property related offenses like grand theft and theft from motor vehicles demonstrated slightly negative correlations, suggesting that unemployment alone does not strongly drive crime rates. This analysis is valuable because it highlights the complexity of social behavior - crime trends cannot be explained by a single factor like employment status. Instead, the findings underscore the importance of considering additional influences such as demographics, enforcement policies, and community conditions when developing strategies to reduce crime.
