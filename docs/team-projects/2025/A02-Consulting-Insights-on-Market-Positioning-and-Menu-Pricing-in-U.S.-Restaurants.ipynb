<span style="font-size: 20px;">***3.4 Data Cleaning***</span>

````python
restaurant_order_clean = restaurant_order.copy()

# Check missing values before cleaning
print("Missing values BEFORE cleaning:\n", restaurant_order.isna().sum())

restaurant_order_clean.columns = [c.lower().strip().replace(" ", "_") for c in restaurant_order_clean.columns]

#Mapping the price ($: Inexpensive, $$: Moderately expensive, $$$: Expensive, $$$$: Very Expensive)
price_map = {'$': 1, '$$': 2, '$$$': 3, '$$$$': 4}
restaurant_order_clean['price_range'] = restaurant_order_clean['price_range'].map(price_map)

#Extractling city and state from full address
def extract_city_state(full_address):
    try:
        parts = [p.strip() for p in str(full_address).split(",")]
        if len(parts) >= 3:
            city = parts[-3]
            state_zip = parts[-2]
            state = state_zip.split()[-1] if state_zip.strip() else ""
            return city, state
        elif len(parts) == 2:
            city = parts[0]
            state = parts[1].split()[-1] if parts[1] else ""
            return city, state
        else:
            return (np.nan, np.nan)
    except:
        return (np.nan, np.nan)

city_state = restaurant_order_clean['full_address'].apply(extract_city_state)
restaurant_order_clean['city'] = city_state.apply(lambda x: x[0])
restaurant_order_clean['state'] = city_state.apply(lambda x: x[1])

restaurant_order_clean['primary_cuisine'] = restaurant_order_clean['category'].fillna("Unknown").apply(lambda x: str(x).split(",")[0].strip())

restaurant_order_clean['ratings'] = pd.to_numeric(restaurant_order_clean['ratings'], errors='coerce')

#Droping unncessary columns
columns_to_drop = ['lat', 'lng', 'price_range_num', 'full_address']
restaurant_order_clean = restaurant_order_clean.drop(columns=[col for col in columns_to_drop if col in restaurant_order_clean.columns])

#Printing
print("\nMissing values AFTER cleaning:\n", restaurant_order_clean.isna().sum())
print("After cleaning - shape:", restaurant_order_clean.shape)
print("Unique states:", restaurant_order_clean['state'].nunique())
print("Unique cuisines (primary):", restaurant_order_clean['primary_cuisine'].nunique())
````

````python
# Check missing values before cleaning
print("Missing values BEFORE cleaning:\n", restaurant_menu.isna().sum())

# Clean menu price: remove currency text and convert to float
def parse_price(p):
    try:
        if pd.isna(p):
            return np.nan
        s = str(p).strip()
        s = s.replace(',', '')
        s = s.replace('USD','').replace('usd','').strip()
        s = s.replace('$','').strip()
        if '-' in s and not s.replace('-', '').strip().isdigit():
            s = s.split('-')[0].strip()
        parts = s.split()
        for token in parts:
            try:
                val = float(token)
                return val
            except:
                continue
        return float(s)
    except:
        return np.nan

restaurant_menu['price'] = restaurant_menu['price'].apply(parse_price)

restaurant_menu = restaurant_menu.dropna(subset=['restaurant_id'])
````

````python
#Common preprocessing for graphs 4.1 - 4.13

top_cuisines = restaurant_order_clean['primary_cuisine'].value_counts().nlargest(15)
top_cuisine_list = top_cuisines.index.tolist()

#Ratings vs Price summary
ratings_price = (
    restaurant_order_clean.dropna(subset=['price_range', 'ratings'])
    .groupby('price_range')['ratings']
    .agg(['mean','count'])
    .reset_index()
)

#Cuisine diversity (top 15 states)
cuisine_diversity = (
    restaurant_order_clean.groupby('state')['primary_cuisine']
    .nunique()
    .sort_values(ascending=False)
    .head(15)
)

#Ratings distribution for top 5 cuisines
top5 = top_cuisines.index[:5].tolist()
ratings_by_cuisine = {
    c: restaurant_order_clean.loc[restaurant_order_clean['primary_cuisine'] == c, 'ratings'].dropna()
    for c in top5
}

#State × cuisine prep
state_restaurant_counts = (
    restaurant_order_clean.groupby("state")["restaurant_id"]
    .nunique()
    .reset_index(name="restaurant_count")
    .sort_values("restaurant_count", ascending=False)
)
states_top5 = state_restaurant_counts.head(5)

state_cuisine_counts = (
    restaurant_order_clean[restaurant_order_clean["state"].isin(states_top5["state"])]
    .groupby(["state", "primary_cuisine"])["restaurant_id"]
    .nunique()
    .reset_index(name="restaurant_count")
)

top5_cuisines_per_state = (
    state_cuisine_counts.sort_values(["state", "restaurant_count"], ascending=[True, False])
    .groupby("state")
    .head(5)
    .reset_index(drop=True)
)

pivot_stacked = (
    top5_cuisines_per_state.pivot_table(
        index="state",
        columns="primary_cuisine",
        values="restaurant_count",
        fill_value=0
    )
)

#Top 12 Cuisines
top12_cuisines = (
    restaurant_order_clean["primary_cuisine"]
    .value_counts()
    .head(12)
    .index
)

#Merging menu
menu_merged = restaurant_menu_clean.merge(restaurant_order_clean, on="restaurant_id", how="left")
menu_top12 = menu_merged[menu_merged["primary_cuisine"].isin(top12_cuisines)]

# Compute average price & rating per cuisine
cuisine_value = (
    restaurant_order_clean.groupby('primary_cuisine')
    .agg(avg_price=('price_range', 'mean'), avg_rating=('ratings', 'mean'))
    .reset_index()
)

top10_cuisines = cuisine_value.nlargest(10, 'avg_price')

# Compute correlation matrix
corr_features = restaurant_order_clean[['price_range', 'ratings', 'restaurant_id']].copy()
corr_features['cuisine_diversity'] = restaurant_order_clean.groupby('state')['primary_cuisine'].transform('nunique')

corr = corr_features.corr()
````

<span style="font-size: 20px;">**4.4 Top 15 Most Common Cuisines Across Restaurants**</span>


Question: What are the most common cuisines across all restaurants, and how significant is their share?

````python
def plot_cuisine_distribution(df, column="primary_cuisine", top_n=15):
    cuisine_counts = df[column].value_counts().head(top_n)
    cuisine_perc = (cuisine_counts / cuisine_counts.sum()) * 100

    plot_df = pd.DataFrame({
        "Cuisine": cuisine_counts.index,
        "Count": cuisine_counts.values,
        "Percentage": cuisine_perc.values
    })

    plt.figure(figsize=(12,6))
    sns.barplot(data=plot_df, x="Count", y="Cuisine", palette='viridis')

    for i, (count, perc) in enumerate(zip(plot_df["Count"], plot_df["Percentage"])):
        plt.text(count+200, i, f"{perc:.1f}%", va="center")

    plt.title(f"Top {top_n} Most Common Cuisines")
    plt.xlabel("Number of Restaurants")
    plt.ylabel("Cuisine")
    plt.show()

    return plot_df

plot_cuisine_distribution(restaurant_order_clean, column="primary_cuisine", top_n=15)
````

**Analysis:**
From this chart, The American cuisine is by far the most common, making up almost 30% of all restaurants. After that, Mexican food, burgers and pizza also take large shares, and together the top four categories already account for more than half of the market. This shows that dining in the U.S. is very concentrated in just a few popular cuisines.

At the same time, there are many smaller categories like breakfast and brunch, desserts, Asian, Chinese, Italian, and Indian food. Each of them only takes a few percent, but they add variety and reflect cultural diversity. Even smaller ones like seafood and bakery still play a role.

<span style="font-size: 20px;">**4.15 State-level Rating Residuals (Actual - Predicted Ratings)**</span>


Question: How do actual restaurant ratings differ from the ratings predicted by economic and pricing factors across U.S. states?

````python
#Prepare model data
df = merged_zip.copy()
df["score"] = pd.to_numeric(df["score"], errors="coerce")
df["price_range"] = df["price_range"].replace({"$":1,"$$":2,"$$$":3,"$$$$":4})
df["price_range"] = pd.to_numeric(df["price_range"], errors="coerce")
df["median_household_income"] = pd.to_numeric(df["median_household_income"], errors="coerce")

model_data = df.dropna(subset=["score","price_range","median_household_income","state"]).copy()

#Regression
X = sm.add_constant(model_data[["median_household_income","price_range"]])
y = model_data["score"]
ols_model = sm.OLS(y, X).fit()
model_data["residual"] = y - ols_model.predict(X)

#Aggregate residuals
state_residuals = (model_data.groupby("state", as_index=False).agg(avg_residual=("residual","mean")).sort_values("avg_residual", ascending=False))

fig = px.bar(state_residuals, x="state", y="avg_residual", color="avg_residual", color_continuous_scale="RdBu_r",
    range_color=[-abs(state_residuals["avg_residual"]).max(), abs(state_residuals["avg_residual"]).max()],
    text=state_residuals["avg_residual"].round(2),title="State-Level Restaurant Ratings Residuals (Actual − Predicted)")

fig.add_hline(y=0, line_dash="dash", line_color="black")
fig.update_traces(textposition="outside", marker_line_color="black", marker_line_width=0.4, hovertemplate="<b>%{x}</b><br>Residual: %{y:.2f}<extra></extra>")

fig.update_layout(plot_bgcolor="white", paper_bgcolor="white", width=1200, height=600, bargap=0.3,
    title=dict(x=0.5, font=dict(size=20)), xaxis=dict(title="State"), yaxis=dict(title="Residual (Higher = More Generous Ratings)"),
    coloraxis_colorbar=dict(title="Residual<br>(Higher = More Generous)"))

fig.show()
````

**Analysis:** This bar chart shows the state-level rating residuals, the difference between the actual restaurant ratings and the predicted values. Louisiana currently ranks first, and its positive residual value has significantly increased, indicating that customers in this state have rated the restaurants much higher than expected.

This suggests that the rating culture in this state is more enthusiastic or lenient. Vermont and Illinois also have ratings higher than expected, indicating that customer feedback is generally positive or lenient.

On the other hand, Idaho, Tennessee, and Indiana have the largest negative residuals, suggesting that diners in these states tend to be more demanding or picky than the model predicted. West Virginia and Ohio also tend to have stricter evaluations.

Overall, this chart highlights the significant differences in the rating behaviors of various states. Cultural concepts and local dining expectations may influence customers' perception and evaluation of the dining experience.

# <span style="font-size: 30px;">**7 Challenges**</span>

While interpreting the data, we encountered several methodological and data-related constraints, the recognition of which is critical for interpreting the strategic recommendations and for guiding future research efforts.


**1.The Missing Ratings Problem -a Data Bias Risk**: Our biggest data risk was that roughly 23% of restaurants had no rating data. When we analyze value versus ratings, we had to ignore these unrated businesses. If those missing restaurants were mostly low-quality (and thus unrated), their exclusion could make our 'Value Matrix' look artificially positive for affordable categories. This could potentially bias our finding on perceived value.


**2.Homogeneity Risk** - Cuisine categorization is too simple and it underestimating the true diversity of the market. This may lead to incorrectly assigning competitors, making our market share and niche analysis less precise than it could be.

**3.No Sales Data** - The dataset lacked essential transactional data like sales volume, order frequency, or basket size. This is a critical limitation because we can observe pricing power (what restaurants charge), but we cannot definitively calculate the price elasticity of demand or confirm if our pricing recommendations will actually lead to higher profits.
