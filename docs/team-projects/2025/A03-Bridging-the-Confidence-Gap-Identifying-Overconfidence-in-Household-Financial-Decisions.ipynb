##**a. Project Proposal**
####**Problem Definition:**
Households often misjudge their financial literacy, either overestimating or underestimating their abilities. This mismatch can lead to poor financial decisions (especially for overcofident households), such as excessive debt, inadequate savings, or limited participation in wealth-building opportunities. The challenge lies in identifying the key drivers of these mismatches and understanding how demographic characteristics, financial attitudes, and behaviors contribute to the gap. By addressing this problem, policymakers, educators, and financial institutions can design more targeted interventions to bridge the gap between financial confidence and capability.

####**Motivation:**
Many households believe they understand finances well enough to make good decisions, but data often reveals a different story. Overconfidence in financial knowledge can quietly lead to costly mistakes, from under-saving to risky borrowing. This project explores how this “data literacy gap” affects real financial behaviors and outcomes. Understanding where and why overconfidence arises can help policymakers, educators, and financial institutions design interventions that empower individuals to make more informed and resilient financial choices.

####**Data Source: Survey of Consumer Finances**

The Survey of Consumer Finances (SCF) is a nationally representative survey of U.S. households conducted every three years by the Federal Reserve Board, with support from the Department of the Treasury. It is considered one of the most authoritative sources of micro-level data on household finances. The SCF provides detailed information on income, net worth, balance sheet components (such as assets, debts, and financial products), saving and investment behavior, credit use, and demographic characteristics of families.

The survey is unique because it combines a standard area-probability sample with an oversample of wealthy households, ensuring that both middle-income families and those at the very top of the wealth distribution are adequately represented. This makes the SCF especially powerful for examining financial inequality, wealth distribution, and household-level financial behavior.

In the context of this project, the SCF (2022) provides the right scope and level of detail to study how households’ financial characteristics and behaviors relate to our initial research question. By leveraging variables on income, debt, savings, financial literacy, and attitudes toward risk, along with the provided survey weights, we can generate results that are both statistically robust and nationally representative. In other words, the SCF allows us not only to analyze the financial health of individual respondents, but also to draw conclusions that generalize to the broader U.S. population.

Link: https://www.federalreserve.gov/econres/scfindex.htm

##**b. Executive Summary**
This report examines the phenomenon of financial overconfidence—households that rate their financial knowledge highly but exhibit low financial health. Using data from the Survey of Consumer Finances (SCF), we identify that approximately 9.1 million U.S. households fall into this overconfident segment, with an additional 27.3 million at risk of becoming financially unhealthy.

Our analysis reveals that overconfident households face persistent challenges in long-term wealth building, debt management, and saving discipline. Despite perceiving themselves as financially knowledgeable, their behaviors indicate structural weaknesses in planning and execution. Importantly, we find that households with an established financial plan, regardless of their literacy level, consistently demonstrate stronger financial health.

These insights suggest that policy interventions and financial institutions can leverage perceived literacy as an entry point for engagement. By offering tailored, “advanced-feeling” planning tools and strategies, stakeholders can help overconfident households translate their confidence into sustainable financial stability, improving both individual resilience and overall economic wellbeing.

##**1b. Recoding survey data**
The SCF data is coded to avoid typos when capturing the answers of each question. In order to correctly identify categorical variables a recoded is needed. To do so we created a 'codebook' using the methodology and instructions from SCF website, and then recoded the dataframe to correctly identify categorical variables.

```python
#Preparing the codebook for mapping
code_map=codebook[['Variable','Code','Label']]
to_recode = code_map["Variable"].unique()  #Extracting the column names of the variables to recode

#Recode relevant variables, since the data came from a survey
id_vars = [col for col in SCFP_filtered.columns if col not in to_recode]                  #Identifying the variables that don't need recoding
long_df = SCFP_filtered.melt(id_vars=id_vars, var_name="Variable", value_name="Code")     #Changing to a long format for easier mapping with the codebook
merged = long_df.merge(code_map, on=["Variable", "Code"], how="left")                     #Merging the codebook and the survey dataset
merged["Label"] = merged["Label"].fillna(merged["Code"])                                  #Filling missing recoded values with the original ones, for those variables that don't requier recoding
recoded = merged.pivot(index=id_vars, columns="Variable", values="Label").reset_index()   #Pivot back to a wide format
recoded.head(3)
```


##**1c. Handling Missing Values**
The SCF does not have missign values since it follows a methodology of imputation to avoid missing data and enhancing the external validity of the data to estimate poblational metrics. Missing data in the survey have been imputed five times using a multiple imputation technique. The information is stored in five separate imputation replicates (implicates). Thus, for the 4,602 families interviewed for the survey, there are 23,010 records in the data set. Seven observations were deleted for the public version of the data set for purposes of disclosure avoidance; thus, there are 22,975 records in the public data set for 4,595 families.


##**1d. Questions' current definition updating**
There are a couple of variables that have a way of encoding that changed over the years and have deprecated values, that if used, should be handled to reflect the current definition of the questions. The adjusted variables were "RACE" and "KNOWL". The first to match the current encoding where "Asian /Pacific Islander" is included into the "Other" category. The second to set the minimum possible value for the self-assessed financial literacy to 0; some people reported not even having zero literacy, for those cases the encoding was -1, but for the purposes of the development of this project -1 and 0 will be considered the same, the lowest possible value of self-assessed financial literacy.

```python
#Updating variables due to changes in its definition from past surveys
recoded["RACE"] = recoded["RACE"].replace(4, 'Other')   #In past surveys 4 was used to Asian/Pacific Islander, now they are grouped in 5 'Other'
recoded["KNOWL"] = recoded["KNOWL"].replace(-1, 0)      #-1 was defined as the lower reported knowledge, not too different from 0, which is going to be the lowest
display(recoded['RACE'].describe(),recoded['KNOWL'].describe())
```

##**1e. Creating a "Financial Health Index"**
To determine if a household has 'healthy' financial habits an index was defined taking into account 4 dimensions:

*   **Liquidity:** To assess short-term capacity to sort financial challenges.
*   **Debt Management:** To assess if their current debt is affordable and sustainable.
*   **Assets Building:** To assess if the household is accumulating wealth over time.

*  **Risk Profile:** to assess household's tendency to take financial risk.

It provides a quick scoring system to measure healthy financial habits and also group segments of the population with similar behavior. We also defined a scale of literacy, low, moderate and high, to group the households according to their self-assesed financial literacy. It is going to be instrumental to identify overconfident households and their financial decisions.


```python
#Defining a function to calculate the financial_health_score
def weighted_median(values, weights):
    values = np.array(values)
    weights = np.array(weights)
    mask = ~np.isnan(values) & ~np.isnan(weights)
    values = values[mask]
    weights = weights[mask]
    if len(values) == 0:
        return np.nan
    sorter = np.argsort(values)
    values = values[sorter]
    weights = weights[sorter]
    cum_weights = np.cumsum(weights)
    cutoff = np.sum(weights) / 2.0
    return values[cum_weights >= cutoff][0]

net_median = weighted_median(recoded['NETWORTH'], recoded['WGT'])

def financial_health_score(row,net_median):
    score = 0
    score += 1 if row["HLIQ"] == 'yes' else 0
    score += 1 if row["HOUSECL"] == 'yes' else 0
    score += 1 if row["BNKRUPLAST5"] == 'no' else 0
    score += 1 if row["FORECLLAST5"] == 'no' else 0
    score += 1 if (row["TURNDOWN"] == 'no' and row["TURNFEAR"] == 'no') else 0
    score += 1 if row["WSAVED"] == 'spend less than income' else 0
    score += 1 if row["HSTOCKS"] == 'yes' else 0
    score += 1 if row["SAVRES6"] == 'yes' else 0
    score += 1 if row["HDEBT"] == 'no' else 0
    score += 1 if row["YESFINRISK"] == 'yes' else 0
    score += 1 if row["NETWORTH"] > net_median else 0
    score += 1 if (row["DEBT2INC"] < 0.4) else 0
    return score

#Calculating the financial_health_score
recoded['financial_index'] = recoded.apply(lambda row: financial_health_score(row, net_median), axis=1)

#Stablishing the literacy perception scale by defining a function
def literacy(x):
    if x <= 3:
        return 'low literacy'
    elif x <= 7:
        return 'moderate literacy'
    else:
        return 'high literacy'

#Assigning the literacy scale
recoded['literacy_scale'] = recoded['KNOWL'].apply(literacy)

#Stablishing the literacy perception scale by defining a function
def health(x):
    if x <= 3:
        return 'low financial health'
    elif x <= 7:
        return 'moderate financial health'
    else:
        return 'high financial health'


#Assigning the health scale
recoded['health_scale'] = recoded['financial_index'].apply(health)
recoded.head(3)
```

###**Relation between self-assesed literacy and financial health**

An initial statement needs to be done,Is self-assessed financial literacy a meaningful predictor of objective financial health?

```python
df = recoded.copy()

# weighted
results = []
for k in sorted(df["KNOWL"].unique()):
    g = df[df["KNOWL"] == k]
    if g.empty:
        continue
    weighted_knowl = (g["KNOWL"] * g["WGT"]).sum() / g["WGT"].sum()
    weighted_fin = (g["financial_index"] * g["WGT"]).sum() / g["WGT"].sum()
    results.append((weighted_knowl, weighted_fin))

weighted_df = pd.DataFrame(results, columns=["KNOWL_weighted", "Financial_index_weighted"])

fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(
    weighted_df["KNOWL_weighted"],
    weighted_df["Financial_index_weighted"],
    marker="o", color="steelblue", lw=2
)

ax.set_xlabel("Self-assessed Literacy")
ax.set_ylabel("Financial Index")
ax.set_title("Three Stages of Financial Health Growth by Literacy Level")
ax.set_ylim(0, 12)
ax.set_xlim(0, 10.5)

# add shade
ax.axvspan(0, 2, color="mistyrose", alpha=0.4, label="0–2: Volatile (peak at 1, trough at 2)\n→ 'Knowledge trap'")
ax.axvspan(2, 7, color="aliceblue", alpha=0.4, label="2–7: Linear growth")
ax.axvspan(7, 10, color="honeydew", alpha=0.4, label="7–10: Accelerated gains")

# keypoint (2, 7, 10)
key_knowl_values = [2, 7, 10]
key_points = []

for val in key_knowl_values:
    idx = (weighted_df["KNOWL_weighted"] - val).abs().idxmin()
    knowl = weighted_df.loc[idx, "KNOWL_weighted"]
    fin = weighted_df.loc[idx, "Financial_index_weighted"]
    key_points.append((knowl, fin))


    ax.scatter(knowl, fin, color="crimson", s=100, zorder=5, edgecolor="black", linewidth=1.2)

    # annotate
    ax.annotate(
        f"KNOWL={knowl:.1f}\nIndex={fin:.2f}",
        xy=(knowl, fin),
        xytext=(knowl + 0.6, fin + 0.6),
        fontsize=9, color="crimson",
        ha="left", va="bottom",
        arrowprops=dict(arrowstyle="->", color="gray", lw=1),
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="lightgray", alpha=0.8)
    )


ax.legend(loc="upper left", frameon=True, framealpha=0.8)

plt.tight_layout()
plt.show()


print("Key points (Weighted values):")
display(pd.DataFrame(key_points, columns=["KNOWL_weighted", "Financial_index_weighted"]))
```

**Findings:**

We discovered three stages:

* 0-2: Volatile (peak at 1, trough at 2) - "knowledge trap"

* 2-7: Gradual linear growth

* 7-10: Accelerated gains


Low self-assesed literacy =<2 is a "danger zone" minimal knowledge leads to overconfidence and poor decisions, worse than complete ignorance (Dunning-Kruger effect). Shallow financial education may harm. Programs must achieve sufficient depth to move learners past the trap threshold.

There is a positive but weak relationship between self-assesed literacy and financial health, which suggest that there are other factors to consider to bridge the gap.

###**Age and lifestage**

Approaching the financial health drivers demographics may have an important impact on financial behaviour and decision making.

Is financial health consistent among different life stages and age groups?

```python
# Defining a function to calculate the metrics needed for the boxplot
def weighted_quantile(values, weights, quantiles):
    sorter = np.argsort(values)
    values, weights = values[sorter], weights[sorter]
    cumsum = np.cumsum(weights)
    cumsum /= cumsum[-1]
    return np.interp(quantiles, cumsum, values)

# Setting a custom order for the age groups
age_order = ['<35','35-44','45-54','65-74','>=75']
recoded["AGECL"] = pd.Categorical(recoded["AGECL"], categories=age_order, ordered=True)

# Calculating the metrics using the function
summary = []
for age in age_order:
    df = recoded[recoded["AGECL"] == age]
    if df.empty:
        continue

    vals = df["financial_index"].to_numpy()
    w = df["WGT"].to_numpy()

    q1, median, q3 = weighted_quantile(vals, w, [0.25, 0.5, 0.75])
    iqr = q3 - q1

    lower = max(vals.min(), q1 - 1.5 * iqr)
    upper = min(vals.max(), q3 + 1.5 * iqr)

    summary.append({
        "AGECL": age,
        "q1": q1, "median": median, "q3": q3,
        "lower": lower, "upper": upper
    })

summary_df = pd.DataFrame(summary)

# Plotting
fig, ax = plt.subplots(figsize=(8,6))

positions = range(len(summary_df))
for pos, row in zip(positions, summary_df.itertuples()):
    # Box
    ax.plot([pos, pos], [row.q1, row.q3], color="black", linewidth=6)
    # Median
    ax.plot(pos, row.median, "ro")
    # Whiskers
    ax.plot([pos, pos], [row.lower, row.q1], color="black", linestyle="--")
    ax.plot([pos, pos], [row.q3, row.upper], color="black", linestyle="--")
    # Whisker caps
    ax.plot([pos-0.1, pos+0.1], [row.lower, row.lower], color="black")
    ax.plot([pos-0.1, pos+0.1], [row.upper, row.upper], color="black")

ax.text(
    0.5, -0.12,
    "Note: Red dots indicate the weighted median financial health score per age group.",
    transform=ax.transAxes,
    ha='center', va='top',
    fontsize=9, color='dimgray', style='italic'
)

ax.set_xticks(positions)
ax.set_xticklabels(summary_df["AGECL"])
ax.set_title("Distribution of Financial Health Score by Age Group")
ax.set_xlabel("Age Group")
ax.set_ylabel("Financial Health Score")

plt.tight_layout()
plt.show()
```

**Findings:**
* There is a positive correlation between age and financial health; older people appear to make better financial decisions.

* However, for individuals over 45, the range of average financial health widens considerably. This suggests that while being older is, on average, correlated with better financial health, this age group contains households with both excellent and poor financial habits. A deeper analysis is needed to identify the specific factors that drive some individuals to maintain healthier habits than others in this age bracket.

##**2b. Overconfident Households' Financial Behaviour and Painpoints**

Using the definition of Overconfident Households its possible to identify their painpoints. By understanding what is causing low financial health levels an action plan to then improve it can be designed.

```python
# Define groups
group = df[(df["KNOWL"] > df["KNOWL"].median()) &
           (df["financial_index"] < df["financial_index"].median())].copy()

rest = df.drop(group.index).copy()

# Define pain point conditions
pain_points = {
    "Low liquidity": lambda d: d["HLIQ"].eq('no'),
    "No home equity or housing credit": lambda d: d["HOUSECL"].eq('no'),
    "Recent bankruptcy": lambda d: d["BNKRUPLAST5"].eq('yes'),
    "Recent foreclosure": lambda d: d["FORECLLAST5"].eq('yes'),
    "Credit turned down or fear of rejection": lambda d: (d["TURNDOWN"].eq('yes') | d["TURNFEAR"].eq('yes')),
    "Doesn't save from income": lambda d: ~d["WSAVED"].eq('spend less than income'),
    "No stock investments": lambda d: d["HSTOCKS"].eq('no'),
    "No savings reserve (6 months)": lambda d: d["SAVRES6"].eq('no'),
    "Has debt": lambda d: d["HDEBT"].eq('yes'),
    "Avoids financial risk": lambda d: d["YESFINRISK"].eq('no'),
    "Below-median net worth": lambda d: d["NETWORTH"] <= net_median,
    "High debt-to-income ratio": lambda d: d["DEBT2INC"] >= 0.4
}

# Weighted share function
def weighted_share(df, condition):
    total_w = df["WGT"].sum()
    return df.loc[condition(df), "WGT"].sum() / total_w if total_w > 0 else 0

# Compute shares and relative differences
pain_data = []
for name, cond in pain_points.items():
    group_share = weighted_share(group, cond)
    rest_share = weighted_share(rest, cond)
    rel_diff = ((group_share - rest_share) / rest_share) * 100 if rest_share > 0 else 0
    pain_data.append((name, group_share, rel_diff))

# Create DataFrame
pain_df = pd.DataFrame(pain_data, columns=["Pain Point", "Weighted Share", "Relative Difference (%)"])
pain_df = pain_df.sort_values("Weighted Share", ascending=True)

# Plot
plt.figure(figsize=(10,6))
bars = plt.barh(pain_df["Pain Point"], pain_df["Weighted Share"], color='#1f77b4')  # blue bars

# Add percentage + relative difference labels
for i, bar in enumerate(bars):
    width = bar.get_width()
    rel_diff = pain_df.iloc[i]["Relative Difference (%)"]
    diff_label = f"(+{rel_diff:.0f}%)" if rel_diff > 0 else f"({rel_diff:.0f}%)"
    plt.text(width + 0.005,
             bar.get_y() + bar.get_height()/2,
             f"{width*100:.1f}% {diff_label}",
             va='center', fontsize=9)

plt.title("Most Common Financial Pain Points", fontsize=13)
plt.xlabel("Share of Group (Weighted)")
plt.tight_layout()
plt.show()
```

**Findings:**
By analyzing the components of the financial health index, the main pain points for overconfident households were identified. Being risk-averse is typical in this group and is consistent with the behavior observed in other households. The three main pain points are:

**1.Savings:** Overconfident households are less likely to save, and this behavior is 20% more common compared to non-overconfident households.

**2.Stocks:** Overconfident households are less likely to invest in stocks, which likely reflects a pattern of not building long-term wealth.

**3. Debt:** Overconfident households tend to have high levels of debt, and this behavior is 42% more common compared to non-overconfident households.

To address these three main pain points, a deeper exploration of the underlying behaviors would help develop more effective improvement strategies.

####**Debt Composition**
Their current debt composition may reveal the reasons for obtaining loans and help identify the most effective repayment plans to reduce overall debt levels.

```python
debt_cols = ['PLOAN1', 'PLOAN2', 'PLOAN3', 'PLOAN4', 'PLOAN5', 'PLOAN6', 'PLOAN7']

legend_map = {
    'PLOAN1': 'Home purchase, cottage or vacation property',
    'PLOAN2': 'Home Improvement',
    'PLOAN3': 'Vehicle Loan',
    'PLOAN4': 'Goods and Services',
    'PLOAN5': 'Mortgage',
    'PLOAN6': 'Student Loan',
    'PLOAN7': 'Others'
}

def weighted_mean(df, cols, weight_col='WGT'):
    w = df[weight_col]
    return pd.Series({col: (df[col] * w).sum() / w.sum() if w.sum() > 0 else 0 for col in cols})

group_mean = weighted_mean(group, debt_cols).to_frame().T
rest_mean  = weighted_mean(rest, debt_cols).to_frame().T
group_mean['Category'] = 'Overconfident Households'
rest_mean['Category']  = 'Non-overconfident households'
composition = pd.concat([group_mean, rest_mean], ignore_index=True)
composition = composition[['Category'] + debt_cols]

# Convert to percentage of total debt per group
composition_pct = composition.copy()
composition_pct[debt_cols] = composition_pct[debt_cols].div(composition_pct[debt_cols].sum(axis=1), axis=0) * 100

# Prepare plot column names
plot_debt_cols = [legend_map[col] for col in debt_cols]
composition_pct_plot = composition_pct.copy()
composition_pct_plot.columns = ['Category'] + plot_debt_cols

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pastel_palette = sns.color_palette("pastel", n_colors=len(plot_debt_cols))

ax = composition_pct_plot.set_index('Category').plot(
    kind='bar',
    stacked=True,
    figsize=(10,6),
    color=pastel_palette
)

plt.ylabel('Debt Composition (%)')
plt.title('Debt Composition by Loan Type: Group vs. Rest')
plt.xticks(rotation=0)
plt.legend(title='Loan Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Add percentage data labels
for i, category in enumerate(composition_pct_plot['Category']):
    cumulative = 0
    for j, col in enumerate(plot_debt_cols):
        percent = composition_pct_plot.loc[i, col]
        if percent > 0:
            bottom = cumulative
            cumulative += percent
            center = bottom + percent / 2
            ax.text(
                i, center, f"{percent:.1f}%",
                ha='center', va='center', color='black', fontsize=9,
                fontweight='bold'
            )


plt.show()
```

##**2c. Having a Financial Plan**

How do financial literacy and financial health jointly influence the likelihood of having a financial plan?

* Are overconfident individuals (high literacy but low financial health) more or less likely to have a financial plan compared to others?
* Does having a financial plan mitigate the negative effects of overconfidence on financial health outcomes?


```python
# Copy and clean data
df_copy = recoded.copy()

# Normalize relevant columns
for col in ['IFINPLAN', 'health_scale', 'literacy_scale']:
    df_copy[col] = df_copy[col].astype(str).str.strip().str.lower()

# Define category orders
literacy_order = ['high literacy', 'moderate literacy', 'low literacy']
health_order = ['high financial health', 'moderate financial health', 'low financial health']

# Calculate weighted percentage of IFINPLAN == 'yes'
weighted = (
    df_copy.groupby(['literacy_scale', 'health_scale'], observed=False)
    .apply(lambda x: (x.loc[x['IFINPLAN'] == 'yes', 'WGT'].sum() / x['WGT'].sum()) * 100)
    .reset_index(name='percent_yes_IFINPLAN')
)

# Ensure all literacy × health combos exist
weighted = (
    weighted.set_index(['literacy_scale', 'health_scale'])
    .reindex(pd.MultiIndex.from_product([literacy_order, health_order],
                                        names=['literacy_scale', 'health_scale']))
    .reset_index()
)
weighted['percent_yes_IFINPLAN'] = weighted['percent_yes_IFINPLAN'].fillna(0)

# Plot weighted grouped bar chart
plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=weighted,
    x='literacy_scale',
    y='percent_yes_IFINPLAN',
    hue='health_scale',
    palette='viridis',
    order=literacy_order,
    hue_order=health_order
)

# Add % labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='edge', padding=2, fontsize=9)

# Customize chart
plt.ylim(0, 100)
plt.title('Weighted % of Respondents with a Financial Plan\nby Literacy Level and Financial Health', fontsize=13)
plt.xlabel('Financial Literacy Level', fontsize=11)
plt.ylabel('Weighted Percentage with Financial Plan (%)', fontsize=11)
plt.legend(title='Financial Health')
plt.tight_layout()
plt.show()
```

2. Overconfident individuals (high literacy, low health) show a low planning rate (~23%), suggesting that confidence without financial stability doesn’t translate into proactive planning.

3. Across all literacy levels, those who plan tend to have better financial health, implying that planning helps close the confidence gap.

#**5. Limitations and Furture Research Suggestion** #

While this study leverages the nationally representative 2022 Survey of Consumer Finances (SCF) to explore financial literacy mismatch, several limitations should be acknowledged.

First, many key variables, including financial literacy, risk attitudes, and perceived financial knowledge, are self-reported and may be influenced by social desirability, recall bias, and question framing effects. The cross-sectional nature of the SCF prevents causal inference, allowing only correlational insights.  

Another important limitation concerns unobserved heterogeneity, such as psychological traits, cultural attitudes toward money, or prior financial socialization experiences, may also influence both confidence and actual financial behavior. These latent variables, which are not captured in the SCF, could therefore bias the estimated mismatch between confidence and actual financial health. A related methodological limitation involves the construction of the Financial Health Index (FHI).

The index aggregates multiple indicators of financial well-being, but the process necessarily involves subjective design choices regarding variable selection and weighting. Different specifications of the FHI could lead to alternative interpretations of financial health and potentially alter the observed strength of relationships within the analysis.

Future research could address these limitations by incorporating longitudinal data to track changes in financial confidence and behavior, employing experimental or improved survey designs (e.g., randomized vignettes or behavioral tasks) to reduce reporting and framing bias, and conducting sensitivity analyses to assess the impact of alternative FHI constructions.

Despite these limitations, the study provides valuable empirical evidence on the confidence gap in household financial behavior and offers a solid foundation for future behavioral and policy-oriented research.
