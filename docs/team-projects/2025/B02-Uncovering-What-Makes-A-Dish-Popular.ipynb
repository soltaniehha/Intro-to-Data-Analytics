**Step 2:** After finding out the exploratory data information, we converted "AggregatedRating" and "ReviewCount" columns into numeric. For the rows that could not be converted into numbers, we treated them as NaN. In the second part of conversion, we deleted any such NaN rows and only kept those recipes that had a reviewcount of more than 1 so that we are able to run a more robust analysis.

```python
#Converting Aggregated Rating and Review Count inti numeric
recipes["AggregatedRating"] = pd.to_numeric(recipes["AggregatedRating"], errors="coerce")
recipes["ReviewCount"] = pd.to_numeric(recipes["ReviewCount"], errors="coerce")
#Filtering so that rows with more than 1 rreview count remain
recipes = recipes.dropna(subset=["AggregatedRating", "ReviewCount"])
recipes = recipes[recipes["ReviewCount"] > 1]
```

**Step 3:** In the next part of our data cleaning, we removed any columns that were not relevant to our problem statement. Columns like Images, RecipeServings, Description were removed.

```python
#removing unnecessary columns according to our problem statement
# remove unnecessary columns
recipes = recipes.drop(columns=["Images", "CookTime", "PrepTime", "RecipeYield", "RecipeInstructions", "Keywords", "RecipeIngredientQuantities", "RecipeServings"])
recipes.head(3)
```

**Step 4:** The "TotalTime" column was essential for our problem statement. Many questions could be answered using this column, but it was in a format that we could not work on. Hence we extracted total minutes from this column. We defined a function called "parse duration" and used a regular expression pattern to find hours and minutes in the string, and save them into a new column "TotalTimeMinutes" to show the result of the conversion. After this conversion was successfully completed, we deleted the original "TotalTime" column.

```python
#fixing Total_Time Column
import re
import pandas as pd

# function to parse ISO 8601 duration strings like "PT24H45M"
def parse_duration(duration):
    if pd.isna(duration):
        return None

    # regex to extract hours and minutes
    pattern = r"PT(?:(\d+)H)?(?:(\d+)M)?"
    match = re.match(pattern, duration)

    if not match:
        return None

    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0

    return hours * 60 + minutes

# apply to TotalTime column
recipes["TotalTimeMinutes"] = recipes["TotalTime"].apply(parse_duration)
recipes[["TotalTime", "TotalTimeMinutes"]].head(3)
```

```python
#removing TotalTime and description column
recipes = recipes.drop(columns=["TotalTime"])
```

```python
recipes = recipes.drop(columns=["Description"])
```

```python
recipes.head(3)
```

**Step 5:** For our 5th Data Cleaning Step, we standardized several important text columns within the recipes dataset. We converted all of them to lowerspace and removed any whitespace. After that, we correctly classified several columns into numeric datatype.

```python
#making the strings uniform
str_cols = ["Name", "AuthorName", "RecipeCategory", "RecipeIngredientParts"]
for col in str_cols:
    recipes[col] = recipes[col].str.lower().str.strip()
```

```python
#classifying numeric columns correctly
num_cols = ["AggregatedRating","ReviewCount","Calories","FatContent","SaturatedFatContent","CholesterolContent","SodiumContent","CarbohydrateContent","FiberContent","SugarContent","ProteinContent"]
for col in num_cols:
    recipes[col] = pd.to_numeric(recipes[col], errors="coerce")
```

**Step 6:** In order to fill in the missing numeric values, we filled each of them with the respecitive mean of their category. In order to do that, we first groupedby the recipes dataset and found the mean of each category. Then, we filled the missing values with the mean. For Example, if FatContent of a recipe in the Dessert category, then that missing value will be filled with the mean of the FatContent of the Dessert category.

```python
# Imputing missing numeric values with category mean
recipes[num_cols] = recipes.groupby("RecipeCategory")[num_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
#checking for missing values
recipes.isna().mean().sort_values(ascending=False)
(recipes.isna().mean() * 100).round(2)
missing_info = recipes.isna().sum().to_frame('Missing Count')
missing_info['Missing %'] = (recipes.isna().mean() * 100).round(2)
print(missing_info)
```

**Step 7:** The next step is to explore the reviews dataset and conduct minimal data cleaning in which we ensured RecipeID is consistent. We also made sure that text columns are clean and we dropped duplicates. Lastly, we aggregated reviews per recipe and kept only those recipes that have reviews, in line with our problem statement.

We created another dataset called recipes_reviews,in which we had merged the recipes columns with the average review rating and the number of reviews from the reviews dataset. This is our final,cleaned dataset that we will use to analyze data and answer our question of "What makes a recipe popular?".

```python
reviews.info()
```

```python
# Minimal cleaning for reviews
reviews_clean = reviews.copy()
# Ensure RecipeId is consistent type
reviews_clean["RecipeId"] = reviews_clean["RecipeId"].astype(str).str.strip()
# Clean text columns
reviews_clean["AuthorName"] = reviews_clean["AuthorName"].str.lower().str.strip()
reviews_clean["Review"] = reviews_clean["Review"].fillna("no review").str.lower().str.strip()
#Dropping duplicates
reviews_clean = reviews_clean.drop_duplicates()
#Aggregating reviews per recipe
reviews_agg = reviews_clean.groupby("RecipeId").agg(avg_review_rating=("Rating", "mean"),num_reviews=("Rating", "count")).reset_index()
# Merging with cleaned recipes ---
recipes_reviews = pd.merge(
    recipes,
    reviews_agg,
    on="RecipeId",
    how="inner"
)
print("Recipes before merge:", len(recipes))
print("Recipes after merge :", len(recipes_reviews))
print(recipes_reviews.head())

```

# Phase 2: Defining the Popularity Criteria
In order to effectively analyze "what makes a recipe popular", we had to set a criteria for what 'popular' actually means. In our second phase of the project, we built a multi-level 'PopularityTier' system of categorization through the integration of two main indicators: user activity (ReviewCount) and satisfaction (AggregatedRating).

By this composite approach, we defined three distinct tags: 'Highly Popular', 'Moderately Popular', and 'Less Popular'.  This new PopularityTier attribute then became a staple of our research, allowing us to delve more deeply and meaningfully into the variables that separate a true success from the rest.

```python
#Checking dataset description to set popularity
recipes_reviews[['AggregatedRating', 'ReviewCount']].describe()

```

```python
import numpy as np

#defining popularity conditions
conditions = [
    (recipes_reviews['ReviewCount'] >= 11) & (recipes_reviews['AggregatedRating'] >= 4.5),
    (recipes_reviews['ReviewCount'].between(5, 10)) & (recipes_reviews['AggregatedRating'] >= 4.3),
    (recipes_reviews['ReviewCount'] < 5) | (recipes_reviews['AggregatedRating'] < 4.3)
]

#Defining labels
choices = ['Highly Popular', 'Moderately Popular', 'Less Popular']

#creating new column
recipes_reviews['PopularityTier'] = np.select(conditions, choices, default='Less Popular')

# Verify distribution
recipes_reviews['PopularityTier'].value_counts()
```

**Analyzing How Nutritional Profiles Affect Recipe Ratings**

```python
import pandas as pd
import numpy as np
# ckleaning the data
subset = recipes_reviews.dropna(subset=["Calories", "ProteinContent", "FiberContent", "avg_review_rating"])
#Removing nutritional values over 1000
subset = subset[
    (subset["Calories"] > 0) & (subset["Calories"] <= 1000) &
    (subset["ProteinContent"] >= 0) & (subset["ProteinContent"] <= 1000) &
    (subset["FiberContent"] >= 0) & (subset["FiberContent"] <= 1000)
]
#creating nutritional level grps
subset["calorie_group"] = pd.qcut(subset["Calories"], q=3, labels=["Low Calorie","Moderate Calorie","High Calorie"])
subset["protein_group"] = pd.qcut(subset["ProteinContent"], q=3, labels=["Low Protein","Moderate Protein","High Protein"])
subset["fiber_group"] = pd.qcut(subset["FiberContent"], q=3, labels=["Low Fiber","Moderate Fiber","High Fiber"])
#Concatenating the nutritional combination labels
subset["nutrition_profile"] = (
    subset["calorie_group"].astype(str) + " / " +
    subset["protein_group"].astype(str) + " / " +
    subset["fiber_group"].astype(str)
)
#calculating average rating and sample count
rating_summary = (
    subset.groupby("nutrition_profile")["avg_review_rating"]
    .agg(["mean","count"])
    .round(2)
    .sort_values("mean", ascending=False)
)
#filtering combinations where 700 < count < 1000
rating_summary_filtered = rating_summary[(rating_summary["count"] > 700) & (rating_summary["count"] < 1000)]

display(rating_summary_filtered)
```

**Takeaway**

The highest-rated nutritional profile identified was "High Calorie / Moderate Protein / Moderate Fiber," which achieved a mean rating of 4.35. The absence of "Low Calorie" and "Low Protein" combinations from the top-rated profiles suggests that users associate higher satisfaction with substantial, energy-rich meals.

Crucially, the top-rated profile avoids extremes, featuring moderate levels of both protein and fiber. This supports the hypothesis that excessive nutritional content can be detrimental to a recipe's appeal. The findings indicate that a well-rounded, satisfying dish is rated more favorably by users than a recipe that is maximized for a single nutritional statistic.

## **3D.Popularity Over Time**


After analyzing the static attributes of recipes, this section introduces the critical dimension of time to understand how popularity trends evolved as the platform grew and matured. We examine this evolution from two key perspectives: the growth in user engagement (reviews and ratings) and the changing volume of content creation (new recipe submissions).

The analysis reveals that as the platform's user base expanded, the average rating metric transformed from a highly volatile number into a stable and trustworthy measure of satisfaction. Simultaneously, the rate of new recipe submissions surged, but this growth was overwhelmingly dominated by content that would ultimately be classified as 'Less Popular'. This temporal context is crucial, as it demonstrates that as the platform grew, it also became an increasingly competitive ecosystem where achieving a high popularity status became a significant challenge.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = recipes_reviews.copy()
#parsing dates
rating_col = "avg_review_rating" if "avg_review_rating" in df.columns else "AggregatedRating"
count_col  = "num_reviews"       if "num_reviews"       in df.columns else "ReviewCount"
df["PublishedDate"] = pd.to_datetime(df.get("DatePublished"), errors="coerce")

#aggregating by month
trend = df.dropna(subset=["PublishedDate"]).groupby(pd.Grouper(key="PublishedDate", freq="M")).agg(avg_rating=(rating_col, "mean"),total_reviews=(count_col, "sum")).reset_index()

#Plotting chart
fig, ax1 = plt.subplots(figsize=(9,4))
ax1.plot(trend["PublishedDate"], trend["avg_rating"], label="Avg Rating")
ax1.set_ylabel("Avg Rating")
ax2 = ax1.twinx()
ax2.plot(trend["PublishedDate"], trend["total_reviews"], label="Total Reviews", ls="--")
ax2.set_ylabel("Total Reviews")
ax1.set_title("Popularity Over Time — Rating & Engagement")
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines+lines2, labels+labels2, loc="upper left")
plt.grid(False)
fig.tight_layout(); plt.show()
```

**Chart 14 Takeaway**

This analysis reveals two distinct trends in the platform's early years. The first is a clear, exponential growth in user engagement, as shown by the 'Total Reviews' (dashed line), which surged from a few thousand to nearly 30,000 between 1999 and mid-2002.

In contrast, the 'Avg Rating' (solid line) was extremely volatile during the initial period (1999-2001), swinging wildly between ratings of 3.8 and a perfect 5.0. This indicates that early ratings were based on a very small number of reviews, making the average highly susceptible to outliers. As the volume of reviews increased significantly after 2001, the average rating stabilized into a much more consistent range of 4.1 to 4.4.

The key takeaway is that the growth in review volume had a stabilizing effect on the average rating. This demonstrates a classic "law of large numbers" principle, where a larger sample size provides a more accurate and trustworthy measure of overall user satisfaction. It suggests that data from the platform's more mature phase is a more reliable indicator of a recipe's true quality and appeal.

Titles and short descriptions that use positive, sensory cues (e.g., “crispy,” “fresh,” “garlic”) attract more engagement and slightly higher ratings, likely because such language sets vivid expectations and helps users “taste” the dish before cooking. Cultural familiarity also matters: our correlation analysis shows overlap in audiences for certain cuisines (e.g., Mediterranean and Asian), suggesting that success rises when flavors feel both recognizable and a little distinctive.
In sum, the most popular recipes balance taste, health, and effort. They rely on familiar ingredients, keep steps manageable, and use inviting language that clarifies why the dish is rewarding. This blend of comfort, practicality, and sensory appeal defines what modern home cooks consider both “worth cooking” and “worth rating highly,” and it provides a clear blueprint for creators and platforms aiming to increase adoption and satisfaction.

Numerical fields required normalization and unit consistency, while text fields often contained irregular formatting or missing strings. We standardized the name of the variable, transformed the list into a string separated by a comma, and extracted useful features like the number of ingredients and the length of the review for quantitative analysis.
