## 1. Introduction

### Problem Statement
Financial literacy plays a critical role in shaping personal and economic well-being: higher levels often enable individuals to make better decisions and achieve greater life satisfaction, while lower levels can lead to poor choices that affect many aspects of life. Building on this understanding, this project evaluates the accuracy of self-assessed financial confidence and financial knowledge by comparing them with individuals’ actual financial literacy scores across the U.S. population. The study further examines how this alignment—or potential discrepancy—varies by key demographic factors, including age, gender, region, and annual income.

### Executive Summary
The goals of our analysis:
- **Define** financial literacy — specifically the key components and factors that make up one’s financial literacy score.
- **Analyze** the existing trends of reported financial confidence and financial knowledge in relation to age, gender, region, and annual income.
- **Evaluate** initial findings from data analysis and visualizations, and compare them with actual financial literacy scores to assess the validity of individuals’ self-assessments.


This report evaluates the accuracy of self-assessed financial confidence and knowledge by comparing them with individuals’ actual financial literacy levels across the U.S. population. Financial literacy has become increasingly essential in light of income disparity and wealth inequality. Initial demographic analysis of survey participants—based on gender, age, region, and income—indicates that males, individuals aged 65+, higher income levels tend to report higher levels of financial confidence and knowledge.

To assess the reliability of self-reported measures, the team developed a Financial Literacy (FinLit) calculator. The tool computes an individual’s financial literacy score using weighted responses from selected survey questions, incorporating variables such as ownership of checking, savings, and retirement accounts, access to emergency funds, health insurance coverage, and other investments. Participants were classified into one of eight categories based on their total scores: Very Low (0-5), Low (6-12), Moderately Low (13-19), Average (20-26), Moderately High (27-33), High (34-39), Very High (40-46), None (0).

Comparison of FinLit scores with self-reported financial confidence shows substantial overlap across all confidence levels, with low-score outliers even among those who “Strongly Agree.” This suggests that confidence alone is not a reliable measure of financial literacy. Similarly, overestimation of financial knowledge is evident among participants who consider themselves to have “Average” knowledge but score lower on the FinLit scale, highlighting a gap between perception and reality. The strongest alignment occurs among individuals with high financial literacy who also report a knowledge score of 6.

While the FinLit score is still exploratory, the study underscores the importance of continuing to refine the tool and survey questions. The findings also highlight the need to expand financial education initiatives across all demographics, through public institutions and workplace programs, to improve financial literacy and overall financial well-being in the United States.

### S1.1: Difference in financial confidence across age group and gender

```python
# Color palette
from matplotlib.colors import LinearSegmentedColormap
sunset_orange_blue = LinearSegmentedColormap.from_list(
    "sunset_orange_blue",
    ["#F26B38", "#F7C99E", "#0067A3"],  # orange → sand → blue
)
```

```python
# --- Step 1: Clean and normalize confidence column ---
# Map qualitative confidence to numeric scale (1–7)
confidence_map = {
    "Strongly disagree": 1,
    "Moderately disagree": 2,
    "Slightly disagree": 3,
    "Neither agree nor disagree": 4,
    "Slightly agree": 5,
    "Moderately agree": 6,
    "Strongly agree": 7
}
df["Financial Confidence (num)"] = df["Financial Confidence"].map(confidence_map)

# Drop missing values for relevant columns
conf_df = df.dropna(subset=["Age Group", "Gender", "Financial Confidence (num)"])

# --- Step 2: Compute mean and standard error for confidence by Age × Gender ---
summary = (
    conf_df.groupby(["Age Group", "Gender"])["Financial Confidence (num)"]
    .agg(["mean", "sem"])
    .reset_index()
)

# --- Step 3: Sort Age Group in logical order ---
age_order = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
summary["Age Group"] = pd.Categorical(summary["Age Group"], categories=age_order, ordered=True)
summary = summary.sort_values("Age Group")

# --- Step 4: Plot line chart with error bars ---
plt.figure(figsize=(9, 6))

# Define custom orange-blue palette (warm vs cool)
gender_palette = {
    "Female": "#F26B38",  # bright orange
    "Male": "#0067A3"     # deep blue
}

sns.lineplot(
    data=summary,
    x="Age Group",
    y="mean",
    hue="Gender",
    palette=gender_palette,
    marker="o",
    linewidth=2,
    err_style="bars",
)

# --- Step 5: Customize chart aesthetics ---
plt.title("Average Financial Confidence by Age and Gender (2009–2024)", fontsize=14, pad=15)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Average Financial Confidence (1–7 scale)", fontsize=12)
plt.legend(title="Gender", loc="upper left")
plt.grid(alpha=0.3)
sns.despine()

plt.tight_layout()
plt.show()
```

Line graph above shows that:
- Financial confidence increases steadily with age for both men and women. The youngest group (18–24) reports the lowest average confidence (around 5.3 on a 7-point scale), while confidence rises consistently across age groups and peaks among those aged 65 and older (above 6.2).

- Across all age groups, men generally show slightly higher financial confidence than women, though the gap remains small and consistent. This suggests that while experience and age play a stronger role in shaping financial confidence, a minor gender gap persists throughout all age levels.

### S1.2: Analyzing financial confidence across regions with age and gender group

```python
# --- Step 1: Clean and normalize confidence column ---
# Map qualitative confidence responses to numeric scale (1–7)
confidence_map = {
    "Strongly disagree": 1,
    "Moderately disagree": 2,
    "Slightly disagree": 3,
    "Neither agree nor disagree": 4,
    "Slightly agree": 5,
    "Moderately agree": 6,
    "Strongly agree": 7
}
df["Financial Confidence (num)"] = df["Financial Confidence"].map(confidence_map)

# Drop missing values for relevant columns
conf_region = df.dropna(subset=["Census Division", "Age Group", "Gender", "Financial Confidence (num)"])

# --- Step 2: Prepare ordered categories for plotting ---
age_order = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
div_order = [
    "New England",
    "Middle Atlantic",
    "East North Central",
    "West North Central",
    "South Atlantic",
    "East South Central",
    "West South Central",
    "Mountain",
    "Pacific"
]
gender_order = ["Female", "Male"]

conf_region["Age Group"] = pd.Categorical(conf_region["Age Group"], categories=age_order, ordered=True)
conf_region["Census Division"] = pd.Categorical(conf_region["Census Division"], categories=div_order, ordered=True)
conf_region["Gender"] = pd.Categorical(conf_region["Gender"], categories=gender_order, ordered=True)

# Combine age and gender into one column (use str.cat to avoid categorical + string error)
short_gender = conf_region["Gender"].map({"Female": "F", "Male": "M"})
conf_region["Age × Gender"] = (
    conf_region["Age Group"].astype("string").str.cat(short_gender.astype("string"), sep=" • ")
)

# --- Step 3: Compute mean financial confidence by Division × (Age × Gender) ---
heat_df = (
    conf_region.groupby(["Census Division", "Age × Gender"])["Financial Confidence (num)"]
    .mean()
    .round(2)
    .reset_index()
)

# Arrange columns in logical order
age_gender_cols = []
for a in age_order:
    for g in ["F", "M"]:
        age_gender_cols.append(f"{a} • {g}")

heat = (
    heat_df.pivot(index="Census Division", columns="Age × Gender", values="Financial Confidence (num)")
    .reindex(index=div_order, columns=age_gender_cols)
)

# --- Step 4: Plot heatmap ---
plt.figure(figsize=(14, 7))
sns.heatmap(
    heat,
    cmap="Oranges",
    annot=True,
    fmt=".2f",
    linewidths=0.3,
    cbar_kws={"label": "Average Financial Confidence (1–7)"}
)

# --- Step 5: Customize chart aesthetics ---
plt.title("Financial Confidence by Division × (Age × Gender) — 2009-2024", fontsize=14, pad=15)
plt.xlabel("Age × Gender", fontsize=12)
plt.ylabel("Census Division", fontsize=12)
plt.xticks(rotation=35, ha="right")
plt.yticks(rotation=0)
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

Above heatmap shows that:
- Financial confidence consistently increases with age across all U.S. regions for both genders, starting around 5.2 among 18–24-year-olds and rising to above 6.2 for adults 65 and older. Men report slightly higher confidence than women within each age group, though the gender gap remains small.

- Across regions, differences are modest but observable. The Middle Atlantic, Mountain, and West North Central divisions show the highest overall confidence, while the East South Central and West South Central divisions tend to lag slightly behind. Overall, age is the strongest factor, with regional and gender variations playing smaller roles.

### S2.1: Analyzing the relationship between financial confidence and saving-spending behavior

```python
# --- Step 1: Prepare clean subset (keep rows with key variables present) ---
save_df = df.copy()

# Fix typo in Likert responses and clean whitespace
save_df["Financial Confidence"] = (
    save_df["Financial Confidence"]
    .replace({"Strongerly disagree": "Strongly disagree"})
    .astype(str).str.strip()
)

# --- Step 2: Normalize binary behavior columns to 0/1 safely ---
# Handles "Yes/No", "Y/N", "True/False", 1/0, etc.
bin_map = {
    "yes": 1, "y": 1, "true": 1, "1": 1, 1: 1, True: 1,
    "no": 0,  "n": 0, "false": 0, "0": 0, 0: 0, False: 0
}

for col in ["Savings Account", "Emergency Funds", "Overdraw"]:
    save_df[col] = (
        save_df[col]
        .astype(str).str.strip().str.lower()
        .map(bin_map)
        .astype(float)
    )

# Drop rows missing any key values
save_df = save_df.dropna(subset=["Financial Confidence", "Savings Account", "Emergency Funds", "Overdraw"])

# --- Step 3: Set Likert order for the X axis (Disagree → Agree) ---
conf_order = [
    "Strongly disagree", "Moderately disagree", "Slightly disagree",
    "Neither agree nor disagree", "Slightly agree", "Moderately agree", "Strongly agree"
]
# Remove any responses not in our defined Likert scale
save_df = save_df[save_df["Financial Confidence"].isin(conf_order)].copy()

save_df["Financial Confidence"] = pd.Categorical(
    save_df["Financial Confidence"], categories=conf_order, ordered=True
)

# --- Step 4: Aggregate mean shares by confidence level ---
save_summary = (
    save_df.groupby("Financial Confidence", observed=False)[
        ["Savings Account", "Emergency Funds", "Overdraw"]
    ].mean().reindex(conf_order).reset_index()
)

# --- Step 5: Melt into long format for seaborn plotting ---
behavior_bars = ["Savings Account", "Emergency Funds", "Overdraw"]
color_map = {
    "Savings Account": "#F26B38",
    "Emergency Funds": "#F7C99E",
    "Overdraw (Overspending)": "#0067A3"
}

long_df = save_summary.melt(
    id_vars="Financial Confidence",
    value_vars=behavior_bars,
    var_name="Behavior",
    value_name="Share"
)
long_df["Behavior"] = long_df["Behavior"].replace({"Overdraw": "Overdraw (Overspending)"})
behavior_order_disp = ["Savings Account", "Emergency Funds", "Overdraw (Overspending)"]

# --- Step 6: Plot grouped bars ---
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(11, 6))

sns.barplot(
    data=long_df,
    x="Financial Confidence",
    y="Share",
    hue="Behavior",
    hue_order=behavior_order_disp,
    palette=color_map,
    alpha=0.9,
    dodge=True,
    legend=False,
    ax=ax
)

# --- Step 7: Add trend lines ---
sns.pointplot(
    data=long_df,
    x="Financial Confidence",
    y="Share",
    hue="Behavior",
    hue_order=behavior_order_disp,
    palette=color_map,
    dodge=0.4,
    join=True,
    markers="o",
    linestyles="--",
    errorbar=None,
    linewidth=2,
    legend=False,
    zorder=5,
    ax=ax
)

# --- Step 8: Custom legend on the right ---
from matplotlib.patches import Patch
from matplotlib.lines import Line2D

handles = [
    Patch(facecolor=color_map["Savings Account"], edgecolor="none", label="Savings Account"),
    Patch(facecolor=color_map["Emergency Funds"], edgecolor="none", label="Emergency Funds"),
    Patch(facecolor=color_map["Overdraw (Overspending)"], edgecolor="none", label="Overdraw (Overspending)"),
    Line2D([0], [0], color=color_map["Savings Account"], lw=2, linestyle="--", marker="o", label="Savings Account (trend)"),
    Line2D([0], [0], color=color_map["Emergency Funds"], lw=2, linestyle="--", marker="o", label="Emergency Funds (trend)"),
    Line2D([0], [0], color=color_map["Overdraw (Overspending)"], lw=2, linestyle="--", marker="o", label="Overdraw (Overspending) (trend)")
]

ax.legend(
    handles=handles, title="Behavior",
    loc="upper left", bbox_to_anchor=(1.02, 1.00),
    borderaxespad=0., frameon=True, facecolor="white", framealpha=0.9,
    fontsize=10, title_fontsize=11, ncol=1
)
fig.subplots_adjust(right=0.78)

# --- Step 9: Final formatting ---
ax.set_title("Saving and Spending Behavior by Financial Confidence Level", fontsize=14, pad=15)
ax.set_xlabel("Financial Confidence (Disagree → Agree)", fontsize=12)
ax.set_ylabel("Share of Respondents (0–1 scale)", fontsize=12)
ax.set_ylim(0, 1)
ax.tick_params(axis="x", rotation=25)
sns.despine()
plt.show()

```

The trends of the graph above shows that:
- Individuals with higher financial confidence are more likely to save and manage money responsibly. As confidence increases, the share of respondents with savings accounts and emergency funds rises sharply, while the likelihood of overdrawing or overspending decreases.

- This pattern suggests that greater financial confidence is strongly linked to positive financial behaviors, including consistent saving and reduced reliance on credit or overdrafts.

### S2.4: Measuring the relationship between higher financial confidence and financial resilience

```python
# --- Step 1: Clean confidence labels robustly (fix typos, spacing, casing) ---
conf_raw = (
    df["Financial Confidence"]
      .astype(str).str.strip().str.lower()
      .replace({
          "strongerly disagree": "strongly disagree",   # common typo
          "strongly  disagree": "strongly disagree",    # double spaces
          "moderately  agree": "moderately agree"
      })
)

# Map to 1–7 using lower-case keys
conf_map_lc = {
    "strongly disagree": 1,
    "moderately disagree": 2,
    "slightly disagree": 3,
    "neither agree nor disagree": 4,
    "slightly agree": 5,
    "moderately agree": 6,
    "strongly agree": 7
}
df["Financial Confidence (num)"] = conf_raw.map(conf_map_lc)

# --- Quick diagnostics: do we have level 1 after cleaning? ---
print("Raw confidence value counts (cleaned):")
print(conf_raw.value_counts(dropna=False).head(15), "\n")

print("Counts by mapped level 1..7:")
print(df["Financial Confidence (num)"].value_counts(dropna=False).sort_index(), "\n")

# --- Step 2: Clean Yes/No columns safely (strip, lowercase, map) ---
binary_cols = ["Emergency Funds", "Overdraw"]
yn_map = {"yes": 1, "no": 0, "y": 1, "n": 0, 1: 1, 0: 0, True: 1, False: 0}

res_df = df[["Financial Confidence (num)"] + binary_cols].copy()
for col in binary_cols:
    s = res_df[col]
    # normalize text but keep non-strings (numbers/bools) intact
    s_norm = np.where(s.apply(lambda x: isinstance(x, str)),
                      s.astype(str).str.strip().str.lower(),
                      s)
    res_df[col] = pd.Series(s_norm).map(yn_map)  # unknowns -> NaN

# --- Step 3: Aggregate shares by confidence level (1..7) ---
order = [1, 2, 3, 4, 5, 6, 7]
res_df = res_df.dropna(subset=["Financial Confidence (num)"])
res_df["Financial Confidence (num)"] = pd.Categorical(res_df["Financial Confidence (num)"],
                                                      categories=order, ordered=True)

summary = (
    res_df.groupby("Financial Confidence (num)", observed=False)
          .agg({"Emergency Funds": "mean", "Overdraw": "mean"})
          .reindex(order)
          .reset_index()
)

# If a level had respondents but all EF/Overdraw are NaN, means will be NaN; set to 0 to show a bar
summary[["Emergency Funds", "Overdraw"]] = summary[["Emergency Funds", "Overdraw"]].fillna(0)

summary["Emergency Funds %"] = summary["Emergency Funds"] * 100
summary["Overdraw %"]        = summary["Overdraw"]        * 100
summary["Neither %"]         = (100 - summary["Emergency Funds %"] - summary["Overdraw %"]).clip(lower=0)

# --- Step 4: Plot ---
plt.figure(figsize=(10, 6))
x = summary["Financial Confidence (num)"].astype(int)

bar1 = plt.bar(x, summary["Overdraw %"], color="#F26B38", label="Overdraw (Overspend)")
bar2 = plt.bar(x, summary["Neither %"], bottom=summary["Overdraw %"],
               color="#F7C99E", edgecolor="#F7C99E", label="Neither")
bar3 = plt.bar(x, summary["Emergency Funds %"],
               bottom=summary["Overdraw %"] + summary["Neither %"],
               color="#0067A3", label="Has Emergency Fund")

def annotate_segments(bars, values):
    for rect, val in zip(bars, values):
        if np.isfinite(val) and val >= 5:
            cx = rect.get_x() + rect.get_width()/2
            cy = rect.get_y() + rect.get_height()/2
            plt.text(cx, cy, f"{val:.0f}%", ha="center", va="center", fontsize=9, color="#222")

annotate_segments(bar1, summary["Overdraw %"].round(0))
annotate_segments(bar2, summary["Neither %"].round(0))
annotate_segments(bar3, summary["Emergency Funds %"].round(0))

plt.title("Financial Resilience by Financial Confidence Level", fontsize=14, pad=15)
plt.xlabel("Financial Confidence Level (1–7 scale)", fontsize=12)
plt.ylabel("Share of Respondents (%)", fontsize=12)
plt.xticks(ticks=x, labels=order)
plt.ylim(0, 100)
plt.grid(axis="y", alpha=0.3)
sns.despine()
plt.legend(title="Behavior Type", bbox_to_anchor=(1.02, 1.0), loc="upper left",
           frameon=True, facecolor="white", framealpha=0.9)
plt.tight_layout()
plt.show()
```

In contrast, among the most confident individuals, nearly 60% have emergency savings and only 14% overspend.

- This shows that greater financial confidence is strongly associated with better financial preparedness and lower vulnerability to financial shocks.

### S2.5: Measuring individuals' financial behavior with their knowledge and confidence levels across demographic groups

```python
# --- Step 1: Normalize core fields (confidence 1–7, knowledge 1–7) ---

confidence_map = {
    "Strongly disagree": 1,
    "Moderately disagree": 2,
    "Slightly disagree": 3,
    "Neither agree nor disagree": 4,
    "Slightly agree": 5,
    "Moderately agree": 6,
    "Strongly agree": 7
}
df.loc[:, "Financial Confidence"] = (
    df["Financial Confidence"]
      .replace({"Strongerly disagree": "Strongly disagree"})
      .astype(str).str.strip()
)
df.loc[:, "Financial Confidence (num)"] = df["Financial Confidence"].map(confidence_map)
df.loc[:, "Financial Knowledge (num)"] = pd.to_numeric(df["Financial Knowledge"], errors="coerce")

# --- Step 2: Robust binary mapping (Yes/No → 1/0) ---
bin_map = {"Yes": 1, "No": 0, "Y": 1, "N": 0, 1: 1, 0: 0, True: 1, False: 0}
for col in ["Emergency Funds", "Overdraw", "Other Investments", "Other Retirement Accounts"]:
    df.loc[:, f"{col} (bin)"] = df[col].map(bin_map).astype("float")  # unknowns -> NaN

# --- Step 3: Build per-row behavior metrics and keep needed columns ---
age_order = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]
metrics_df = (
    df[[
        "Gender", "Age Group",
        "Financial Knowledge (num)", "Financial Confidence (num)",
        "Other Investments (bin)", "Other Retirement Accounts (bin)",
        "Emergency Funds (bin)", "Overdraw (bin)"
    ]]
    .dropna(subset=["Gender", "Age Group", "Financial Knowledge (num)", "Financial Confidence (num)"])
    .copy()
)

# InvestmentRate = mean of (has other investments, has other retirement accounts)
metrics_df.loc[:, "InvestmentRate"] = metrics_df[["Other Investments (bin)", "Other Retirement Accounts (bin)"]].mean(axis=1)

# Resilience = has emergency fund AND not overdrawing (treat missing overdraft as no penalty)
metrics_df.loc[:, "Resilience"] = metrics_df["Emergency Funds (bin)"] * (1 - metrics_df["Overdraw (bin)"]).fillna(1.0)

# --- Step 4: Aggregate metrics by Gender × Age Group ---
grouped = (
    metrics_df
    .groupby(["Gender", "Age Group"], observed=False)
    .agg({
        "Financial Knowledge (num)": "mean",
        "Financial Confidence (num)": "mean",
        "InvestmentRate": "mean",
        "Resilience": "mean"
    })
    .reset_index()
)

# Order age groups and sort
grouped.loc[:, "Age Group"] = pd.Categorical(grouped["Age Group"], categories=age_order, ordered=True)
grouped = grouped.sort_values(["Gender", "Age Group"])

# --- Step 5: Global min–max scale (across both genders) ---
radar_cols_raw = [
    "Financial Knowledge (num)",
    "Financial Confidence (num)",
    "InvestmentRate",
    "Resilience",
]
radar_labels = ["Knowledge", "Confidence", "InvestmentRate", "Resilience"]

for col in radar_cols_raw:
    mn, mx = grouped[col].min(), grouped[col].max()
    scale = (mx - mn) if (mx - mn) != 0 else 1.0
    grouped.loc[:, f"{col}__scaled"] = ((grouped[col] - mn) / scale).clip(0, 1)

radar_cols_scaled = [c + "__scaled" for c in radar_cols_raw]

# --- Step 6: Helper for radar angles ---
def radar_angles(n_vars: int):
    theta = np.linspace(0, 2*np.pi, n_vars, endpoint=False)
    return np.concatenate([theta, [theta[0]]])  # close the polygon

angles = radar_angles(len(radar_cols_scaled))

# =========================================================
# --- Step 7: Plot radar (facet by Gender, overlay Age groups; gender-branded shades) ---

# Brand colors
gender_brand = {"Female": "#F26B38", "Male": "#0067A3"}  # orange / blue

def make_shades(base_hex: str, n: int, start=0.22, end=0.98):
    """
    Create n shades from very light to the base color.
    start/end in [0,1]: 0 is white, 1 is base color.
    """
    base = np.array(to_rgb(base_hex))
    levels = np.linspace(start, end, n)
    return [tuple((1 - t) * 1.0 + t * base) for t in levels]

# Keep panel order consistent if both genders are present
preferred_order = ["Female", "Male"]
genders_present = grouped["Gender"].dropna().unique().tolist()
genders = [g for g in preferred_order if g in genders_present] or genders_present

sns.set_style("white")
fig, axes = plt.subplots(
    1, len(genders),
    subplot_kw={"projection": "polar"},
    figsize=(6.0 * len(genders), 5.8)
)

if len(genders) == 1:
    axes = [axes]

# Collect handles for separate Female/Male age legends
female_handles, male_handles = [], []

for ax, g in zip(axes, genders):
    sub = grouped[grouped["Gender"] == g]

    # Light → dark shades of the gender color for each age group
    shades = make_shades(gender_brand.get(g, "#333333"), n=len(age_order), start=0.22, end=0.98)

    # Draw one polygon per age group
    for i, ag in enumerate(age_order):
        row = sub[sub["Age Group"] == ag]
        if row.empty:
            continue
        vals = row[radar_cols_scaled].iloc[0].values
        vals = np.concatenate([vals, [vals[0]]])  # close polygon
        ln, = ax.plot(angles, vals, color=shades[i], linewidth=2, label=ag)
        ax.fill(angles, vals, color=shades[i], alpha=0.15)

        # stash a handle per gender for legends
        if g == "Female":
            female_handles.append(Line2D([0],[0], color=shades[i], lw=2, label=ag))
        elif g == "Male":
            male_handles.append(Line2D([0],[0], color=shades[i], lw=2, label=ag))

    # Axis styling
    ax.set_yticks([0.25, 0.5, 0.75, 1.0])
    ax.set_yticklabels(["0.25", "0.50", "0.75", "1.00"], fontsize=9)
    ax.set_ylim(0, 1)
    ax.grid(alpha=0.25)

    # ---- Custom x tick labels: push "InvestmentRate" & "Knowledge" outward ----
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels([])  # we will place our own texts
    for k, lab in enumerate(radar_labels):
        # more room for right/left labels
        r_offset = 1.10 if lab in ("InvestmentRate", "Knowledge") else 1.03
        ax.text(angles[k], r_offset, lab, ha="center", va="center", fontsize=11)

    # --- Gender label ABOVE each radar ---
    ax.text(
        0.5, 1.20, g,
        transform=ax.transAxes,
        ha="center", va="center",
        fontsize=14, fontweight="bold",
        color=gender_brand.get(g, "#333333")
    )

# --- Global title ---
fig.suptitle(
    "Radar of Knowledge, Confidence, Investment, Resilience — by Gender × Age (global min–max scaled)",
    fontsize=14, y=0.95
)

# --- Tiny gender legend at the top ---
fig.legend(
    [Line2D([0],[0], color=gender_brand["Female"], lw=3),
     Line2D([0],[0], color=gender_brand["Male"],   lw=3)],
    ["Female", "Male"],
    loc="upper center", bbox_to_anchor=(0.5, 1.02), ncol=2, frameon=False
)

# --- Separate age-group legends for Female and Male (so both are visible) ---
if female_handles:
    fig.legend(
        female_handles, [h.get_label() for h in female_handles],
        title="Age Group — Female",
        loc="lower left", bbox_to_anchor=(0.06, -0.03),
        ncol=len(age_order),
        frameon=True, framealpha=0.95,
        fontsize=8, title_fontsize=9, handlelength=1.8, handletextpad=0.4
    )

if male_handles:
    fig.legend(
        male_handles, [h.get_label() for h in male_handles],
        title="Age Group — Male",
        loc="lower right", bbox_to_anchor=(0.94, -0.03),
        ncol=len(age_order),
        frameon=True, framealpha=0.95,
        fontsize=8, title_fontsize=9, handlelength=1.8, handletextpad=0.4
    )

# Layout: bring panels slightly inward & give more center room
plt.tight_layout(rect=[0, 0.08, 1, 0.96])
plt.subplots_adjust(left=0.06, right=0.94, wspace=0.45)
plt.show()

```

The chart aboves shows that:
- Financial behaviors generally align with individuals’ knowledge and confidence levels, but the degree of alignment varies by gender and age. For both men and women, older groups show stronger balance across confidence, knowledge, investment, and resilience, indicating greater financial stability with experience.

- However, men exhibit more evenly developed profiles, with confidence, knowledge, and investment closely aligned. Women’s profiles show relatively higher confidence than investment or resilience, especially in younger groups, suggesting that while confidence grows, translating it into consistent financial behaviors takes more time and experience.

### S3.3: Indentifying regions with the strongest overall financial health
In this analysis, we combined confidence, knowledge, savings, credit, and emergency preparedness to calculate financial health.

```python
# --- Step 1: Prepare & clean columns ---
cols = [
    "Census Division",
    "Financial Confidence",
    "Financial Knowledge",
    "Savings Account",
    "Overdraw",
    "Credit Card Balance Interest",
    "Emergency Funds",
]
d = df[cols].copy()

# Likert (confidence) → numeric 1–7 (robust lowercase map)
conf_raw = (
    d["Financial Confidence"].astype(str).str.strip().str.lower()
      .replace({"strongerly disagree": "strongly disagree"})  # common typo
)
likert_map_lc = {
    "strongly disagree": 1, "moderately disagree": 2, "slightly disagree": 3,
    "neither agree nor disagree": 4, "slightly agree": 5,
    "moderately agree": 6, "strongly agree": 7
}
d["Financial Confidence (num)"] = conf_raw.map(likert_map_lc)

# Knowledge → numeric (1–7)
d["Financial Knowledge (num)"] = pd.to_numeric(d["Financial Knowledge"], errors="coerce")

# Helper: binarize Yes/No safely (others → NaN)
def binarize_yes_no(s: pd.Series) -> pd.Series:
    yn_map = {"yes": 1, "no": 0, "y": 1, "n": 0, 1: 1, 0: 0, True: 1, False: 0}
    s2 = np.where(s.apply(lambda x: isinstance(x, str)), s.astype(str).str.strip().str.lower(), s)
    return pd.Series(s2).map(yn_map).astype("float")

d["Savings Account (bin)"] = binarize_yes_no(d["Savings Account"])
d["Overdraw (bin)"] = binarize_yes_no(d["Overdraw"])
d["Credit Card Balance Interest (bin)"] = binarize_yes_no(d["Credit Card Balance Interest"])
d["Emergency Funds (bin)"] = binarize_yes_no(d["Emergency Funds"])

# Drop rows missing any required component or division
needed = [
    "Census Division", "Financial Confidence (num)", "Financial Knowledge (num)",
    "Savings Account (bin)", "Overdraw (bin)", "Credit Card Balance Interest (bin)", "Emergency Funds (bin)"
]
d = d.dropna(subset=needed)

# --- Step 2: Build component scores ---
d["Savings (score)"] = d["Savings Account (bin)"]
d["Credit (score)"] = (1 - d["Overdraw (bin)"] + 1 - d["Credit Card Balance Interest (bin)"]) / 2.0
d["Emergency (score)"] = d["Emergency Funds (bin)"]

# --- Step 3: Aggregate by region, normalize each component (0–1), build composite (0–100) ---
region_means = (
    d.groupby("Census Division", as_index=False)[[
        "Financial Confidence (num)", "Financial Knowledge (num)",
        "Savings (score)", "Credit (score)", "Emergency (score)"
    ]].mean()
)

from sklearn.preprocessing import MinMaxScaler
to_scale = [
    "Financial Confidence (num)", "Financial Knowledge (num)",
    "Savings (score)", "Credit (score)", "Emergency (score)"
]
scaler = MinMaxScaler()
scaled = scaler.fit_transform(region_means[to_scale])
scaled_df = pd.DataFrame(scaled, columns=[c + " (scaled)" for c in to_scale])

region_health = pd.concat([region_means[["Census Division"]], scaled_df], axis=1)
region_health["Composite Score"] = scaled_df.mean(axis=1) * 100
region_health = region_health.sort_values("Composite Score", ascending=False).reset_index(drop=True)

# --- Step 3.5 (debug): make sure we have data
print("Rows by division:", len(region_health))
print(region_health[["Census Division", "Composite Score"]].head(), "\n")

# --- Step 4: Visualize ranking (horizontal bar chart) ---
sns.set(style="whitegrid", font_scale=1.1)
fig, ax = plt.subplots(figsize=(10, 6))

# Convert your colormap into a list of N colors (one per bar)
n = len(region_health)
palette = [sunset_orange_blue(i / max(1, n - 1)) for i in range(n)]

sns.barplot(
    data=region_health,
    y="Census Division", x="Composite Score",
    palette=palette, ax=ax
)

# Annotate bars
for i, v in enumerate(region_health["Composite Score"]):
    ax.text(v + 0.8, i, f"{v:.1f}", va="center", fontsize=9)

# --- Step 5: Titles & labels ---
ax.set_title("Comprehensive Financial Health by U.S. Region", fontsize=14, pad=12)
ax.set_xlabel("Composite Financial Health Index (0–100)")
ax.set_ylabel("")
sns.despine()
plt.tight_layout()
plt.show()

# --- Optional: tidy table with raw means + composite ---
pretty = (
    region_means
      .merge(region_health[["Census Division", "Composite Score"]], on="Census Division")
      .rename(columns={
          "Financial Confidence (num)": "Confidence",
          "Financial Knowledge (num)": "Knowledge",
          "Savings (score)": "Savings",
          "Credit (score)": "Credit",
          "Emergency (score)": "Emergency"
      })
      .sort_values("Composite Score", ascending=False)
      .round(2)
)
display(pretty)
```

Results in the chart and table show that:
- The Middle Atlantic region demonstrates the strongest overall financial health, with the highest composite score of 83.2, supported by strong confidence (6.0), solid financial knowledge (5.23), and high savings and credit indicators. The East North Central (66.4) and South Atlantic (64.0) regions also perform well, showing balanced strengths across knowledge, savings, and emergency preparedness.

- In contrast, the West South Central (6.1) and Mountain (29.6) regions lag far behind, reflecting weaker performance across nearly all dimensions. Overall, regions with higher financial confidence and savings behavior tend to show stronger, more resilient financial health profiles.

- However, the wide score ranges and overlapping boxes across confidence levels suggest considerable variation—many highly confident individuals score no better than moderately confident ones. This means that while confidence provides a general signal of financial literacy, it is not a precise predictor and often overstates actual ability.

### S4.2: Analyzing bias in self-reported financial knowledge using actual financial literacy level

```python
# --- Step 1: Reset & reload dataset (clean start) ---
import matplotlib.pyplot as plt
plt.close("all")  # close any existing figures

raw_df = pd.read_csv("Combined_09_12_24.csv")
df = raw_df.copy()

# --- Step 2: Define category orders ---
finlit_order = ["High", "Moderately High", "Average", "Moderately Low", "Low", "Very Low"]
knowledge_order = ["1", "2", "3", "4", "5", "6", "7", "Don't know", "Prefer not to say"]

# --- Step 3: Build normalized crosstab (Actual FinLit × Self-Reported Knowledge) ---
cross_knowledge = pd.crosstab(
    df["FinLit_Level"],
    df["Financial Knowledge"],
    normalize="index"
)

# Reindex to ensure consistent order (fill missing with zeros)
cross_knowledge = cross_knowledge.reindex(index=finlit_order)
cross_knowledge = cross_knowledge.reindex(columns=knowledge_order, fill_value=0)

# --- Step 4: Define a smooth blue gradient colormap ---
from matplotlib.colors import LinearSegmentedColormap
blue_gradient = LinearSegmentedColormap.from_list(
    "blue_gradient",
    ["#D9EAF7", "#66B2FF", "#004C99"]  # light → medium → dark blue
)

# --- Step 5: Plot heatmap ---
plt.figure(figsize=(11, 6))
ax = sns.heatmap(
    cross_knowledge,
    cmap=blue_gradient,
    annot=True,
    fmt=".1%",
    vmin=0.0, vmax=0.45,
    cbar_kws={'label': 'Percentage'}
)

# --- Step 6: Colorbar formatting ---
cbar = ax.collections[0].colorbar
ticks = [0.0, 0.1, 0.2, 0.3, 0.4, 0.45]
cbar.set_ticks(ticks)
cbar.set_ticklabels([f"{t:.2f}" for t in ticks])

# --- Step 7: Layout polish & labels ---
ax.set_title("Alignment Between Actual Financial Literacy and Self-Reported Knowledge", fontsize=14, pad=15)
ax.set_xlabel("Self-Reported Financial Knowledge (1–7, Don't know, Prefer not to say)", fontsize=12)
ax.set_ylabel("Actual Financial Literacy Level", fontsize=12)
plt.xticks(rotation=30, ha="right")
sns.despine()
plt.tight_layout()
plt.show()
```

The heatmap above explains that:
- Most individuals tend to overestimate their financial knowledge relative to their actual literacy level. The heatmap shows that even among respondents with low or moderately low actual literacy, a large share self-report their knowledge as 5 or 6 on a 7-point scale, suggesting inflated self-perceptions.

- Only a small portion of respondents with truly high literacy accurately identify themselves as highly knowledgeable, indicating that confidence often exceeds competence. This highlights a consistent overconfidence bias, where people believe they know more about financial matters than they actually do.
