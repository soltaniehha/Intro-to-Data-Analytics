# **Executive Summary**

This notebook explores the dynamic relationships between U.S. Treasury yields, equities markets, and macro indicators in the past decade to enhance prediction accuracy. Using time-series methods and visual analytics, it presents a snapshot of interest-rate level, yield-curve shape, and volatility, how they reflect economic cycles and sentiment.
The work recognizes dominant trends such as the post-pandemic spike in yields, the existing yield-curve inversion since 2022, and the shifting bond–equity correlation during inflationary shock episodes. The evidence is that while yield-curve inversions remain excellent recessionary leading indicator signs, their impact on equity performance is regime-sensitive on the basis of prevailing inflation and policy regimes.
<br><br>
Rolling regressions and forecasting diagnostics indicate that market relationships change over time—equities' sensitivity to rate movements and model predictability change abruptly between regimes. The paper finds that use of dynamic, regime-sensitive models and leading indicators provides more powerful predictive information than static correlations.
Briefly, the yield curve continues to be a good leading macro indicator, but interpreting it in a monetary policy, inflation, and market expectations context is the secret to successful forecasting and risk management.

## **Cleaning mainData**

### **Code Block 1**

*drop(index=range(0, 5))* ==> removes the first 5 rows.

*reset_index(drop=True)* ==> resets the row numbers, but in our code it wasn’t saved back, so the old numbering stays.

```python
# To start, we will remove the first five rows because they contain header information rather than actual data values.
mainData_cp = mainData_cp.drop(index=range(0, 5))
mainData_cp.reset_index(drop=True)
mainData_cp.head()
```

### **Code Block 2**

It renames the columns so that bond yield columns become standardized like market_yield_10-year while keeping other columns unchanged.

```python
# We will rename the columns of mainData_cp to a cleaner and more consistent format.
new_columns = ["date"]
for col in mainData_cp.columns[1:]:
    match = re.search(r'(\d+)-(month|year)', col)
    if match:
        new_columns.append(f"market_yield_{match.group(1)}-{match.group(2)}")
    else:
        new_columns.append(col)

mainData_cp.columns = new_columns
mainData_cp.head()
```

### **Code Block 3**

*isnull().sum()* ==> counts how many missing (NaN) values each column has.

*isin(['ND']).sum()* ==> counts how many cells have the string "ND" in each column.

*print()* statements ==> clearly separate the two results when displaying.

***"The Block is used to identifu whether or not the Dataframe has Null Values or not"***

```python
# Then we will begin by checking for any missing values in each column.

# Print the count of missing values (NaN) in each column
print("Missing values:")
print(mainData_cp.isnull().sum())

# Print the count of values explicitly marked as 'ND' in each column
print("\nValues marked as 'ND':")
print(mainData_cp.isin(['ND']).sum())
```

### **Code Block 4**

*replace('ND', pd.NA)* ==> changes all "ND" entries into proper missing values (NaN).

*dropna()* ==> removes all rows that have any missing values.

*reset_index(drop=True)* ==> resets row numbers so the index starts from 0 again.

*head()* ==> shows the first 5 rows of the cleaned dataset.

***It removes holiday rows (with missing or ND values) so only valid trading days remain***

```python
# Since these missing values and values marked as 'ND' correspond to holidays when no market data was published, we decided to drop these values.
# Removing them ensures that the dataset only contains valid trading days, which avoids bias and simplifies subsequent analysis.

# Replace 'ND' with NaN so it can be treated as missing
mainData_cp = mainData_cp.replace('ND', pd.NA)

# Drop all rows that contain any missing values
mainData_cp = mainData_cp.dropna()

# Reset the index after dropping rows so it starts from 0 again
mainData_cp = mainData_cp.reset_index(drop=True)

mainData_cp.head()
```

### **Code Block 5**

*pd.to_datetime(..., errors='coerce')* ==> converts the date column to datetime; if something can’t be parsed, it becomes NaT.

*Loop through columns* ==> for every column except date:

*pd.to_numeric(..., errors='coerce')* ==> converts values to float; if conversion fails, it becomes NaN.

*print(dtypes)* ==> shows the data type of each column after conversion.

***The code ensures the date column is in datetime format and all other columns are numeric (float)***

```python
# We will convert the 'date' column to datetime format and convert other columns to float.

# Convert the 'date' column to datetime format
mainData_cp['date'] = pd.to_datetime(mainData_cp['date'], errors='coerce')

# Convert all columns except 'date' to float
for col in mainData_cp.columns:
    if col != 'date':
        mainData_cp[col] = pd.to_numeric(mainData_cp[col], errors='coerce')

print(mainData_cp.dtypes)
```

## **Statistical Test: Stationarity Check (ADF Test) of S&P 500 and 10-Year Treasury Yield**

**Question:**
Are the S&P 500 and 10-Year Treasury Yield time series stationary, or do they exhibit trends and non-constant variance over time?

**Interpretation:**
The Augmented Dickey-Fuller (ADF) Test is used to determine whether a time series is stationary — meaning its statistical properties (mean, variance, and autocorrelation) remain constant over time. A high p-value (> 0.05) indicates non-stationarity, suggesting the presence of a trend or unit root. In contrast, a low p-value (≤ 0.05) implies stationarity.
The results show that both the S&P 500 and 10-Year Treasury Yield in their original forms are non-stationary, as financial data often trends with time. However, after taking log differences (S&P 500) and first differences (10Y Yield), the p-values fall below 0.05, confirming that the transformed series are stationary.

**Answer:**
Both the S&P 500 and 10-Year Treasury Yield are non-stationary in their raw form, indicating that they contain trends over time. After differencing, both series become stationary, making them suitable for further time series analysis such as correlation, regression, or vector autoregression (VAR) modeling.

```python
# Function to perform the Augmented Dickey-Fuller (ADF) Test for stationarity
def adf_test(series, name=""):
    result = adfuller(series.dropna())
    print(f"ADF Test for {name}")
    print(f"ADF Statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    for key, value in result[4].items():
        print(f"Critical Value ({key}): {value:.4f}")
     # Interpretation based on 5% significance level
    if result[1] <= 0.05:
        print("Stationary (reject null of unit root)\n")
    else:
        print("Non-stationary (cannot reject null)\n")

# Run ADF test on original time series to check for stationarity
adf_test(data["s&p500"], "S&P 500")
adf_test(data["US10Y"], "10-Year Treasury Yield")
```

```python
# Create differenced versions of both series to remove trends and stabilize variance
data_diff = pd.DataFrame({
    "S&P 500_logdiff": np.log(data["s&p500"]).diff(),
    "10-Year Treasury Yield_diff": data["US10Y"].diff()
}).dropna()

# Re-run ADF tests on differenced series to confirm stationarity
adf_test(data_diff["S&P 500_logdiff"], "S&P 500 Log Difference")
adf_test(data_diff["10-Year Treasury Yield_diff"], "DGS10 Difference")
```

Yes—most 2Y “jump days” line up with higher-than-usual VIX readings, showing a clear co-movement between extreme rate shocks and equity fear. The largest jump episodes coincide with well-known stress or policy inflection points, reinforcing the idea that bond-market shocks and equity-market risk perception tend to spike together, even if a few jump days occur without a pronounced VIX surge.

# **Graph 10: S&P 500 Sensitivity to 2Y & 10Y Rate Shocks — 252-Day Rolling Betas**

**Question:**
<br>
How do equities’ sensitivities to short-rate and long-rate shocks evolve over time—i.e., does the S&P 500 react more to 2Y (policy/front-end) moves or 10Y (duration/term-premium) moves—and when do rate moves explain a larger share of daily equity returns?

**Interpretation:**
<br>The plot shows 252-day rolling OLS betas of S&P daily log returns on Δ2Y and Δ10Y (in bps). Negative betas mean stocks tend to fall when yields rise; the larger the magnitude, the stronger the sensitivity. Divergence between the β₂ (2Y) and β₁₀ (10Y) lines indicates which part of the curve is “in charge.” The secondary line is rolling R², signaling how much of equities’ day-to-day variation the two rate shocks jointly explain. Higher R² typically aligns with macro/policy stress when rates are a dominant driver; lower R² implies other forces (earnings, risk sentiment, liquidity) matter more.

**Answer:**
<br>Equity rate sensitivity is clearly time-varying and is **usually stronger to the 2Y** than the 10Y, with β₂ showing larger magnitude and bigger swings—evidence that front-end, policy-driven shocks dominate the equity/rates link in many regimes. Betas can compress or even flip sign briefly around easing/reflation windows, but the prevailing pattern is negative. The rolling **R² spikes during stress/tightening episodes**, indicating rate moves temporarily explain more of daily equity returns; outside those periods, R² recedes and non-rate fundamentals and risk appetite take over.


```python
# Rolling Equity–Rate Betas
sp = SnP500[["Date", "S&P500"]].dropna()
sp = sp.rename(columns={"Date":"date", "S&P500":"sp500"})
sp["date"] = pd.to_datetime(sp["date"])
sp = sp.set_index("date")
sp["r"] = np.log(sp["sp500"] / sp["sp500"].shift(1))

yields = mainData_cp[["date", "market_yield_2-year", "market_yield_10-year"]].dropna()
yields["date"] = pd.to_datetime(yields["date"])
yields = yields.set_index("date")
yields["Δ2Y"] = yields["market_yield_2-year"].diff() * 100
yields["Δ10Y"] = yields["market_yield_10-year"].diff() * 100

merged = sp.join(yields, how="inner").dropna()
window = 252
betas_2Y, betas_10Y, r2s = [], [], []

for i in range(window, len(merged)):
    X = merged.iloc[i-window:i][["Δ2Y", "Δ10Y"]].values
    y = merged.iloc[i-window:i]["r"].values
    X = np.column_stack([np.ones(X.shape[0]), X])
    beta, _, _, _ = np.linalg.lstsq(X, y, rcond=None)
    yhat = X @ beta
    ssr, sst = np.sum((yhat - y.mean())**2), np.sum((y - y.mean())**2)
    betas_2Y.append(beta[1])
    betas_10Y.append(beta[2])
    r2s.append(ssr/sst)

merged = merged.iloc[window:].copy()
merged["B2Y"] = betas_2Y
merged["B10Y"] = betas_10Y
merged["R^2"] = r2s

fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(merged.index, merged["B2Y"], label=r'$\beta_2$ (2Y)', color="blue")
ax1.plot(merged.index, merged["B10Y"], label=r'$\beta_{10}$ (10Y)', color="green")
ax1.set_ylabel("Rolling Beta")
ax1.set_xlabel("Date")
ax1.legend(loc="upper left")
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(merged.index, merged["R^2"], label="Rolling R$^2$", color="orange", alpha=0.5)
ax2.set_ylabel("R$^2$")
plt.title("Rolling Equity-Rate Betas (252-day OLS)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Graph 12**: **Cross-Correlation — Δ(2s10s Slope) vs ΔVIX (Lags −20 to +20 Days)**


**Question:**
Does day-to-day change in the 2s10s yield-curve slope lead changes in equity volatility (VIX), or does VIX tend to move first?

**Interpretation:**
The stem plot shows the correlation between (\Delta \text{slope}*{t+L}) and (\Delta \mathrm{VIX}*{t}) for lags (L\in[-20,+20]). Stems **left** of zero (negative (L)) ask whether the slope moves **before** VIX; stems **right** of zero (positive (L)) ask whether the slope moves **after** VIX. A **negative** value means curve **flattening** ((\Delta\text{slope}\downarrow)) goes with **rising** VIX ((\Delta\mathrm{VIX}\uparrow)). The farther a stem is from zero, the stronger the lead–lag; values near zero indicate little systematic timing.

**Answer:**
The plot shows a **negative contemporaneous/near-term relationship**, with the largest-magnitude stems clustered close to zero lag. This implies that when the curve **flattens**, VIX typically **rises** at the same time or within the next few sessions. Beyond that short window, correlations fade toward zero, so the effect is useful for risk monitoring but not a standalone timing signal.


```python
# Cross-Correlation: ΔSlope vs ΔVIX (lags −20…+20)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# VIX from volitilityIndex_cp (robust column detection)
vix = volitilityIndex_cp.copy()
vix.columns = [c.strip().lower() for c in vix.columns]
date_col = next((c for c in vix.columns if c in ("date", "observation_date") or "date" in c), None)
vix_col  = next((c for c in vix.columns if "vix" in c), None)
if date_col is None or vix_col is None:
    raise ValueError("Could not find date or VIX column in volitilityIndex_cp")

vix = vix[[date_col, vix_col]].rename(columns={date_col: "date", vix_col: "vix"})
vix["date"] = pd.to_datetime(vix["date"], errors="coerce")
vix["vix"]  = pd.to_numeric(vix["vix"], errors="coerce")
vix = vix.dropna().set_index("date").sort_index()

# Treasury 2Y & 10Y from mainData_cp; compute slope
yields = mainData_cp[["date", "market_yield_2-year", "market_yield_10-year"]].copy()
yields["date"] = pd.to_datetime(yields["date"], errors="coerce")
yields = yields.dropna().set_index("date").sort_index()
yields["slope"] = yields["market_yield_10-year"] - yields["market_yield_2-year"]  # pct points

# Merge + first differences
merged = yields.join(vix, how="inner")
merged["Δslope"] = merged["slope"].diff() * 100.0   # bps
merged["ΔVIX"]   = merged["vix"].diff()             # index points
merged = merged.dropna()

# Cross-correlation for lags −20…+20
lags  = np.arange(-20, 21)
corrs = np.array([merged["Δslope"].shift(l).corr(merged["ΔVIX"]) for l in lags])

# Quick table (optional to view in console)
xcorr = pd.DataFrame({"lag": lags, "corr": corrs}).set_index("lag")
print("Top |corr| lags:\n", xcorr.reindex(xcorr["corr"].abs().sort_values(ascending=False).index).head(5))

# Plot (no 'use_line_collection' kwarg)
fig, ax = plt.subplots(figsize=(11, 4.5))
markerline, stemlines, baseline = ax.stem(lags, corrs, basefmt=" ")
plt.setp(stemlines, linewidth=1.4, alpha=0.9)
plt.setp(markerline, markersize=5)
ax.axhline(0, color="black", linestyle="--", linewidth=1)
ax.set_title("Cross-Correlation: ΔSlope vs ΔVIX")
ax.set_xlabel("Lag (days)  [L>0: slope AFTER VIX;  L<0: slope BEFORE VIX]")
ax.set_ylabel("Correlation")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# **Graph 13: Treasury Curve Butterfly Curvature vs Forward 3-Month S&P 500 Return**


**Question:**
<br>Does the shape of the Treasury curve’s **butterfly curvature**— *C_t = 2dot10Y_t - (3M_t+30Y_t)* in bps—help explain the **next-3-month** S&P 500 return?

**Interpretation:**
<br>Each dot pairs today’s curvature (x-axis, bps) with the **forward 3-month annualized** log return of the S&P (y-axis). The red quadratic fit *R_{t->t+63} ~ a+bC_t+cC_t^2* summarizes the central tendency and captures possible **non-linear** effects: if the fit is hump-shaped (inverted-U), it implies the best equity outcomes occur around **moderate positive curvature** (a mildly “humped” curve), while **very negative** curvature (U-shaped curve: tight front/long ends vs mid) or **very high** curvature both associate with weaker forward returns. Dispersion around the fit shows the relationship is noisy—macro, earnings, and risk appetite also matter.

**Answer:**
<br>Yes—based on the quadratic fit, forward S&P returns tend to be **highest at moderate positive curvature** and **weaker at the extremes** (very low/negative or very high curvature). In other words, equities generally fare better when the curve isn’t distorted, while unusually **concave or convex** curve shapes coincide with softer next-quarter equity performance. The effect is meaningful but not deterministic, so curvature is best used as a **conditioning variable** alongside other signals rather than a standalone predictor.


```python
# Butterfly Curvature vs Forward 3M Return
sp = SnP500[["Date", "S&P500"]].dropna()
sp["Date"] = pd.to_datetime(sp["Date"])
sp = sp.set_index("Date")
sp["fwd_3m"] = np.log(sp["S&P500"].shift(-63) / sp["S&P500"]) * 4  # annualized

yields = mainData_cp[["date", "market_yield_3-month", "market_yield_10-year", "market_yield_30-year"]].dropna()
yields["date"] = pd.to_datetime(yields["date"])
yields = yields.set_index("date")
yields["curvature"] = 2*yields["market_yield_10-year"] - (yields["market_yield_3-month"] + yields["market_yield_30-year"])
yields["curvature"] *= 100

merged = sp.join(yields, how="inner").dropna()

plt.figure(figsize=(8,6))
plt.scatter(merged["curvature"], merged["fwd_3m"], alpha=0.4)
fit = np.poly1d(np.polyfit(merged["curvature"], merged["fwd_3m"], 2))
x_vals = np.linspace(merged["curvature"].min(), merged["curvature"].max(), 200)
plt.plot(x_vals, fit(x_vals), color="red", lw=2)
plt.xlabel("Curve Curvature (bps)")
plt.ylabel("Forward 3M Annualized Return")
plt.title("Butterfly Curvature vs Forward S&P Return")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Graph 14**: **S&P 500 Rate-Move Attribution — Rolling R² and Residual Std (252-Day OLS)**


**Question:**
<br>How much of the S&P 500’s day-to-day return variance is explained by front- and long-end rate shocks (Δ2Y, Δ10Y), and when do rates become the dominant driver?

**Interpretation:**
<br>The top panel shows a 252-day rolling R^2 from the OLS model r_t​=α+β_2​Δ2Y_t​+β_10​Δ10Y_t​+ε_t​, where yield changes are in **bps**. Higher (R^2) means rate moves jointly explain more of equity return variation. The bottom panel is the **rolling residual standard deviation** σ_ε, the volatility of returns not captured by rate shocks. Peaks in (R^2) flag **rate-dominant regimes** (policy/tightening, macro shock windows); troughs indicate periods when other forces (earnings, risk appetite, liquidity) matter more. Spikes in σ_ε signal equity-specific stress beyond rates.

**Answer:**
<br>Rates explain a **modest share on average**, but their influence is **episodic**: (R^2) climbs during policy and macro stress—showing equities are temporarily rate-driven—while it falls back in calmer regimes, where non-rate factors dominate and (\sigma_\varepsilon) is relatively higher. In short, the equity–rates link strengthens in tightening/shock phases and weakens otherwise, making rate shocks a **conditional** (not constant) driver of daily S&P returns.


```python
# Rate-Move Attribution: Rolling R² and Residual Std
sp = SnP500[["Date", "S&P500"]].dropna()
sp = sp.rename(columns={"Date": "date", "S&P500": "sp500"})
sp["date"] = pd.to_datetime(sp["date"])
sp = sp.set_index("date")
sp["r"] = np.log(sp["sp500"] / sp["sp500"].shift(1))

yields = mainData_cp[["date", "market_yield_2-year", "market_yield_10-year"]].dropna()
yields["date"] = pd.to_datetime(yields["date"])
yields = yields.set_index("date")
yields["Δ2Y"] = yields["market_yield_2-year"].diff() * 100
yields["Δ10Y"] = yields["market_yield_10-year"].diff() * 100

merged = sp.join(yields, how="inner").dropna()

window = 252
r2s, resid_std = [], []

for i in range(window, len(merged)):
    X = merged.iloc[i-window:i][["Δ2Y", "Δ10Y"]].values
    y = merged.iloc[i-window:i]["r"].values
    X = np.column_stack([np.ones(X.shape[0]), X])
    beta, _, _, _ = np.linalg.lstsq(X, y, rcond=None)
    yhat = X @ beta
    resid = y - yhat
    ssr, sst = np.sum((yhat - y.mean())**2), np.sum((y - y.mean())**2)
    r2s.append(ssr/sst)
    resid_std.append(resid.std())

merged = merged.iloc[window:].copy()
merged["R²_roll"] = r2s
merged["σ_resid"] = resid_std

fig, ax1 = plt.subplots(2, 1, figsize=(12,8), sharex=True)
ax1[0].plot(merged.index, merged["R²_roll"], color="blue", label="Rolling R²")
ax1[0].legend(); ax1[0].grid(True)
ax1[1].plot(merged.index, merged["σ_resid"], color="orange", label="Residual Std (σₑ)")
ax1[1].legend(); ax1[1].grid(True)
plt.suptitle("Rate-Move Attribution of Equities (Rolling OLS)")
plt.xlabel("Years")
plt.tight_layout()
plt.show()
```
